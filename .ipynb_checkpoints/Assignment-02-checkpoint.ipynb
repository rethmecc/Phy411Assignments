## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def above(a=1.):
    N=200000
    x = np.random.normal(0,1,N)
    noise=np.logspace(-2.,2.,40)
    n = np.zeros((N,40))
    for noisei in range(0,40):
        n[:,noisei] = np.random.normal(0,noise[noisei],N)
    y = np.zeros((N,40))
    for i in range (0,40):
        y[:,i] = a*x + n[:,i]

    r = np.zeros((40))
    for i in range(0,40):
        r[i] = np.sum((x-np.mean(x))*(y[:,i]-np.mean(y[:,i])))/(N-1)/(np.std(x)*np.std(y[:,i],ddof=1))#CHECK!!!

    sy=np.zeros((40))
    for i in range(0,40):
        sy[i] = np.std(y[:,i])
    fig,ax=plt.subplots(1,1)
    ax.loglog(sy/noise,r,'.')
    ax.set_xlabel('sy/noise');ax.set_ylabel('rxy');ax.set_title('rxy vs inverse noise (a=%1.2f)'%a)
    ax.set_ylim([0.001,2]);ax.set_xlim([0.5,200])
    print r
above()
```

**2** Repeat the above, but with a different value of $a$.  Comment on how (and why) $a$ changes the value of $r_{xy}$.  

```python
above(0.5)
above(2)
```

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

```python
import scipy.stats as stats
def Q2(N=10,noise=1.0):
    b = np.zeros(10000)
    syx = np.zeros(10000)
    for i in range(0,10000):
        x = np.random.normal(0,1,N)
        a = 1.0
        n = np.random.normal(0,noise,N)
        y = a*x + n

        b[i] = (np.sum(x*y)-N*np.mean(x)*np.mean(y))/(np.sum(x*x)-N*np.mean(x)*np.mean(x))
        syx[i] = np.sqrt(np.sum((y-b[i]*x)**2)/(N-2))

    fig,(ax,bx)=plt.subplots(1,2)
    fig.set_size_inches(10,5)
    ax.hist(b,100,normed=True);
    ax.set_xlabel('b')
    ax.set_ylabel('PDF')

    bx.hist(syx,100,normed=True);
    bx.set_xlabel('syx')
    bx.set_ylabel('PDF')
    print 'syx=%1.3f'%np.mean(syx)
    return b,syx,x,N
b,syx,x,N=Q2()
```

They are both normally distributed, as expected. 

**2** Assuming $s_{y|x}=1.$ (which is the same as our value of $n$), plot the student-t probability density function on the histogram of b and show that they compare favorably (if they don't you might need to check your work!)

<hr>

```python
def student(b,syx,x,N):
    fig,bx=plt.subplots(1,1)
    fig.set_size_inches(10,5)
    bx.hist(b,100,normed=True);
    
    xAxis = np.linspace(-10., 10, 1000)
    sc=np.mean(syx)/np.sqrt(np.sum((x-np.mean(x))**2))
    plt.plot(xAxis, stats.t.pdf(xAxis,df=N-2,loc=1,scale=sc))
    
    bx.set_xlabel('b')
    bx.set_ylabel('PDF')
b,syx,x,N=Q2()
student(b,syx,x,N)
```

**3** Do the same procedure for a different number of data points, and a different value for the noise level $n$.  Remember to scale your theoretical histogram by the new expected value of $s_{y|x}$.

```python
bNew,syxNew,xNew,NNew=Q2(5,2)
student(bNew,syxNew,xNew,NNew)
```

## Q3: Linear comparison of Deep Cove data to James Bay temperature data

```python
# Load the data:

import scipy.stats as stats
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]
```

**1** Make a scatterplot of Plot James Bay versus Deep Cove.

```python
fig,ax=plt.subplots(1,1)
ax.plot(dc,jb,'.')
ax.set_xlabel('Deep Cove Temp.');ax.set_ylabel('James Bay Temp.');ax.set_title('James Bay vs Deep Cove Weather Data')
```

**2** Compute the linear Correlation Co-efficient.

```python
N=np.shape(jb)[0]
r = np.zeros((N))
jbClean = jb[np.isfinite(jb)]
dcClean = dc[np.isfinite(dc)]
integrand=(dc-np.mean(dcClean))*(jb-np.mean(jbClean))
integrand=integrand[np.isfinite(integrand)]
r = np.sum(integrand)/(N-1)/(np.std(dcClean)*np.std(jbClean))
print 'r_dcjb=%1.2f'%r
```

**3** Fit a straight line to the James Bay data as a fucntion of the Deep Cove data.  Comment on the fit.  (Do **NOT** use a canned routine for the fitting)

<hr>

```python
x=dc;y=jb
xy = x*y
xClean=x[np.isfinite(x)*np.isfinite(y)]
yClean=y[np.isfinite(x)*np.isfinite(y)]
N = np.shape(xClean)[0]
b = (np.sum(xClean*yClean)-N*np.mean(xClean)*np.mean(yClean))/(np.sum(xClean**2)-N*np.mean(xClean)**2)
print 'Slope is %1.2f'%b
a = np.mean(yClean)-b*np.mean(xClean)
print 'Intercept is %1.2f'%a
yFit=b*xClean+a

def fitQ3():
    fig,(ax,bx)=plt.subplots(1,2)
    fig.set_size_inches(15,5)

    ax.plot(dc,jb,'.')
    ax.plot(xClean,yFit,'-')
    ax.set_xlabel('Deep Cove Temp.');ax.set_ylabel('James Bay Temp.');ax.set_title('James Bay vs Deep Cove Weather Data')

    (n,xx,yy,im)=bx.hist2d(xClean,yClean,bins=100,cmap=cm.Reds,cmin=0.,normed=True)
    plt.colorbar(im,shrink=0.5)
    bx.set_xlabel('Deep Cove Temp.');bx.set_ylabel('James Bay Temp.');bx.set_title('James Bay vs Deep Cove Weather Data')
    bx.plot(xClean,yFit,'-')
    
    return ax,bx
fitQ3()
```

The line appears to fit the data quite well, as seen in both the colour map and the scatter plot above. The colour plot shows this best, since it shows that the line does in fact go through as much data as possible.

**4** Add the confidence intervals to your fit, and comment.  ***NO CANNED ROUTINES***

```python
syx = np.sqrt(np.sum((yClean-yFit)**2)/(N-2))
print syx

bdn,bup=stats.t.interval(0.95,df=N-2,scale=syx)
bb=np.array([bdn,bup])
print bb
bb=bb/np.sqrt(np.sum((xClean-np.mean(xClean))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])

ax,bx=fitQ3()
xx=np.sort(xClean)
yhathigh=(b+bb[1])*(xx-np.mean(xClean))+np.mean(yClean)
yhatlow=(b+bb[0])*(xx-np.mean(xClean))+np.mean(yClean)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.4)
```

The error bars on this fit are not even visible. I verified that the code is working properly by adding an integer to the bounds and plotting, and as expected I did see the error boundaries. This is probably due to the very large number of the data points.

**5** Note that we have assumed that the Deep Cove data is independent and that the James Bay data is dependent.  There is no reason to assume this!  Do the fit in the other direction, and compare the slopes.

```python
x=jb;y=dc
xy = x*y
xClean=x[np.isfinite(x)*np.isfinite(y)]
yClean=y[np.isfinite(x)*np.isfinite(y)]
N = np.shape(xClean)[0]
b = (np.sum(xClean*yClean)-N*np.mean(xClean)*np.mean(yClean))/(np.sum(xClean**2)-N*np.mean(xClean)**2)
print 'slope is %1.2f'%b
a = np.mean(yClean)-b*np.mean(xClean)
print 'Intercept is %1.2f'%a
yFit=b*xClean+a

fig,(ax,bx)=plt.subplots(1,2)
fig.set_size_inches(15,5)

ax.plot(xClean,yClean,'.')
ax.plot(xClean,yFit,'-')
ax.set_ylabel('Deep Cove Temp.');ax.set_xlabel('James Bay Temp.');ax.set_title('Deep Cove Weather Data vs James Bay')

(n,xx,yy,im)=bx.hist2d(xClean,yClean,bins=100,cmap=cm.Reds,cmin=0.,normed=True)
plt.colorbar(im,shrink=0.5)
bx.set_ylabel('Deep Cove Temp.');bx.set_xlabel('James Bay Temp.');bx.set_title('Deep Cove Weather Data vs James Bay')
bx.plot(xClean,yFit,'-')

syx = np.sqrt(np.sum((yClean-yFit)**2)/(N-2))
print syx

bdn,bup=stats.t.interval(0.95,df=N-2,scale=syx)
bb=np.array([bdn,bup])
print bb
bb=bb/np.sqrt(np.sum((xClean-np.mean(xClean))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])

xx=np.sort(xClean)
yhathigh=(b+bb[1])*(xx-np.mean(xClean))+np.mean(yClean)
yhatlow=(b+bb[0])*(xx-np.mean(xClean))+np.mean(yClean)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.4)
```

The slope in this case is 1.23, while in the previous case it was 0.76. They are not the recipricol of one another due to the non-zero intercept. Both also have extremely small error bounds.

**NOTE** There is a way to treat both axes independently called "Neutral Regression".  It is relatively simple, and involves normalizing the total distance of each point from the trend line. <a href="http://journals.ametsoc.org/doi/pdf/10.1175/1520-0426(1999)016%3C0876%3AAPFANR%3E2.0.CO%3B2">(Marsden 1999)</a>. This technique is strongly preferred if there is no reason to think that one data set is dependent on the other.  A caveat is that the data sets shoudl be normalized before fitting so that the "distance" has meaning.
