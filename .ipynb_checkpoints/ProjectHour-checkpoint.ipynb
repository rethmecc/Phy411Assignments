```python
numStat=np.shape(hour[:,0])[0]
numStat=numStat-1

lon=coast[:len(coast)/2]
lat=coast[len(coast)/2:]

statLoc=hour[:,:2]
x=np.zeros(numStat)
y=np.zeros(numStat)
x[:19]=statLoc[:19,0]-360
x[19:]=statLoc[20:,0]-360
y[:19]=statLoc[:19,1]
y[19:]=statLoc[20:,1]

T=np.shape(hour[0,5:])[0]
time=np.arange(T)
#print hour[0,2:]
start=0
end=T

temp=np.zeros((numStat,T))
temp[0:19,:]=hour[0:19,5:]
temp[19:,:]=hour[20:,5:]
for i in range(numStat):
    
    t=np.where(np.isfinite(temp[i,:]))[0]
    #tt=np.arange(t[0],t[-1]+1,1)
    some=temp[i,t]
    f = interp1d(t, some)
    #temp[i,tt]=f(tt)
    if t[0]>start:
        start=t[0]
    if t[-1]<end:
        end=t[-1]
    ni=0
    for j in range(T):
        if ni==0:
            print np.shape(range(t[-1]-t[0]))
            print np.shape(temp)
            ni=1
        if np.isfinite(temp[i,j])==False:
            vals=np.where(np.isfinite(temp[:,j]))[0]
        #    if i==10 and j>26106 and j<20158:
        #        if ni==0:
        #            print i
        #            ni+=1
        #        print vals
            if np.shape(vals)[0]==0:
                temp[i,j]=f(j)
            else:
                temp[i,j]=np.mean(temp[vals,j])
temp0=temp[:,start:end]-np.mean(temp[:,start:end],0)
```

```python
def plotColor(temp,column,isMode=False,Cp=0,modes=modes,start=0,end=-1,time=np.arange(1)):
    if end<0:
        end=len(temp[0,:])
        if isMode==True:
            end=len(modes[0,:])
    if len(time)<2:
        time=np.arange(len(temp[0,:]))
        if isMode==True:
            time=np.arange(len(modes[0,:]))
    fig,axs=plt.subplots(1,2,figsize=(20,8))
    ax=axs[0]
    ax.plot(lon,lat)
    ax.set_xlabel('Degrees Longitude')
    ax.set_ylabel('Degrees Latitude')
    xi=np.linspace(-124,-123.2,200)
    yi=np.linspace(48.2,49,200)
    #z = np.cos(x*2*np.pi/4.3)*signal.chirp(y+2.,f0=.05,t1=4.,f1=1.)
    #print z
    #print np.shape(temp)
    Xi,Yi=np.meshgrid(xi,yi)
    #print np.shape(x)
    #print np.shape(y)
    Xx = np.vstack((x,y)).T

    #print Xi,Yi
    zi = interpolate.griddata(Xx,temp[:,column],(Xi,Yi),method='cubic')
    #print xi
    #print yi
    #print statLoc
    #print zi[np.isfinite(zi)]

    X,Y=np.meshgrid(xi,yi)

    pcm=ax.pcolormesh(xi,yi,mlab.ma.masked_invalid(zi),cmap=cm.jet)
    if isMode==True:
        pcm.set_clim([-1,1])
    plt.colorbar(pcm,ax=ax,shrink=0.5)
    ax.plot(x,y,'.g',markersize=6)
    #ax.set_xlim(-124,lon[1356])
    ax.set_xlim(-124,-123.2)
    ax.set_ylim(48.2,49.0)
    line=1.*lat
    #print np.where(lon==max(lon))
    line[:1356]=49
    line[1356:]=48.2
    #print len(lon)
    ax.fill_between(lon,lat,line,facecolor=(0.8,0.8,0.8),linewidth=0.0,linestyle='dotted',color='grey')
    ax.fill_between((lon[1356]-0.001,-123.2),48.2,49,facecolor=(0.8,0.8,0.8),color=(0.8,0.8,0.8))
    
    ax.set_aspect(1/np.cos(48.6*np.pi/180.))
    
    if isMode==True:
        times=np.arange(T)
        #print np.shape(time)
        #print np.shape(times[time])
        #print np.shape(modes[column,:])
        if len(time)>10000:
            axs[1].plot_date(times[time]/24.+733408.0+3/24.,modes[column,:],'.')
        else:
            axs[1].plot_date(times[time],modes[column,:])
        axs[1].set_title('Time series from mode %i'%(column+1))
        axs[1].set_xlim(733408.0,733408.0+26074/24.)
        axs[0].set_title('Mode %i includes %1.2f percent of the variance'%((column+1),Cp[column]/np.sum(Cp)*100.))
    else:
        ax.set_title('Temperature distribution ($^\circ$C)')
        axs[1].set_visible(False)
        
    return fig

fig=plotColor(temp,5866)
fig.savefig('ColorSample.pdf')
```

```python
def EOF(start=0,end=len(temp0[0,:])):
    
    time=np.zeros(len(temp0[0,:]),dtype=bool)
    time[start:end]=True
    for i in range(4):
        if start+24*365*i<0:
            time[0:end+24*365*i]=True
        else:
            time[start+24*365*i:end+24*365*i]=True
        
    temp1=temp0[:,time]*1.#-np.mean(temp0[:,time],0)
    times=np.arange(T)
    #dates=np.linspace(start,end,)
    #np.linspace
    #plt.plot_date(time)
    #print len(temp1[1,:])
    #print np.shape(time)
    #temp0=temp[:,start:end]-np.mean(temp[:,start:end])

    C=np.cov(temp1)
    Cp,E=np.linalg.eig(C)

    for i in range(np.shape(E)[1] ):
        ii = np.argmax(np.abs(E[:,i]))
        E[:,i]=E[:,i]/E[ii,i]
    modes=np.dot(np.linalg.inv(E),temp1)

    fig1=plotColor(E,0,True,Cp,modes,start,end,time)
    fig2=plotColor(E,1,True,Cp,modes,start,end,time)
    fig3=plotColor(E,2,True,Cp,modes,start,end,time)
    fig4=plotColor(E,3,True,Cp,modes,start,end,time)
    
    return fig1,fig2,fig3,fig4
fig1,fig2,fig3,fig4=EOF()
fig1.savefig('EOFYearM1.pdf')
fig2.savefig('EOFYearM2.pdf')
fig3.savefig('EOFYearM3.pdf')
fig4.savefig('EOFYearM4.pdf')
```

```python
def spec(x):
    N = np.shape(x)[0]
    t=np.arange(N)
    n = int(np.floor(N/20))
    nfft=256*8
    Gxx = np.zeros((20*10,nfft/2+1))
    num=0
    tbin = np.zeros(20*10)
    for ind in range(0,N-n,int(np.floor(n/10))):
    #    print ind
        px,f=np.squeeze(mlab.psd(x[ind+np.arange(0,n)],nfft,Fs=1/np.median(np.diff(t))))
        Gxx[num,:]=px
        num=num+1
        tbin[num]=np.mean(t[ind+np.arange(0,n)])
    tbin=tbin[0:num]
    Gxx=Gxx[0:num,:]
    fig,ax=plt.subplots(1,1)
    ax.loglog(f,Gxx.transpose())
    plt.xlabel('f [$\mathrm{hr^{-1}}$]');
    plt.ylabel(r'$G_{xx} \mathrm{[m^2 hr]}$')
    
    return tbin,f,Gxx,fig
tbin,f,Gxx,fig=spec(temp[1,:])
fig.savefig('spec.pdf')
```

```python
def cont(tbin,f,Gxx,low=-2,high=1.2):
    fig,ax=plt.subplots(1,1,figsize=(20,8))
    plt.axes(ax)
    #print tbin
    pcm=ax.pcolormesh(tbin/24.+733408.0,f,np.log10(Gxx.transpose()))
    plt.xticks(np.arange(733408, 733408+25415/24.+1, 90.0))
    ax.xaxis.set_major_formatter(dates.DateFormatter('%D'))
    fig.autofmt_xdate()
    plt.xlim((733408,733408+25415/24.))
    #plt.ylim((0,0.5))
    pcm.set_clim((low,high))
    plt.colorbar(pcm)
    plt.title(r'$log_{10}(G_{xx})$')
    #plt.xlabel('t [days]')
    plt.ylabel('F [hr$^{-1}$]')
    return fig
fig=cont(tbin,f,Gxx)
fig.savefig('ContourFull.pdf')
```

```python
def bandPassh(dc,ax,freq=24.,db=100,dbu=0.1):
    fc=2./(24.*60)
    nblocks=16.
    
    wind,fwind=mlab.psd(dc,NFFT=int(len(dc)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(dc)/nblocks/2.))
    ax[0].loglog(fwind,wind,label='original')
    ax[0].set_xlabel('f [1/min]')
    ax[0].set_ylabel('$G_{xx}(f)$')
    ax[0].legend()
    ax[0].grid(True)

    t=np.linspace(0,len(dc)/24.,len(dc))
    ax[1].plot(t,dc,label='unfiltered')

    fNyq=1./2.
    fp = 0.042*24./freq
    fs = 0.24*24./freq
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,dbu, db)
    b,a=signal.ellip(n,dbu,db,fn)
    y = signal.filtfilt(b,a,dc)

    fp = 0.042*24./freq
    fs = 0.42*24./freq
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,dbu, db)
    b, a = signal.ellip(n, dbu, db, fn, 'high')
    y = signal.filtfilt(b,a,y)

    ax[1].plot(t,y,label='filtered')
    ax[1].set_ylabel('Temperature ($^\circ C$)')
    ax[1].set_xlabel('Days')
    ax[1].legend()
    ax[1].set_xlim(0,30/24.*freq)
    ax[1].grid(True)
    wind2,fwind2=mlab.psd(y,NFFT=int(len(y)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(y)/nblocks/2.))
    ax[0].loglog(fwind2,wind2,label='filtered')
    
    return ax,t,y

fig,ax=plt.subplots(4,2)
fig.set_size_inches(15,20)

ax[0,:],t,xF1=bandPassh(temp[1,:]*1.,ax[0,:],24)
ax[1,:],t,xF2=bandPassh(temp[1,:]*1.,ax[1,:],12.,145)
ax[2,:],t,xF3=bandPassh(temp[1,:]*1.,ax[2,:],8)
ax[3,:],t,xF4=bandPassh(temp[1,:]*1.,ax[3,:],6,10)
```

The $\mathrm{\frac{2}{24hr}}$ peak does not show the same variation, but seems fairly constant throughout the year. The $\mathrm{\frac{3}{24hr}}$ peak show the same variation as the $\mathrm{\frac{2}{24hr}}$ peak.

```python
def highPass(dc,ax,freq=24.,db=100,dbu=0.1):
    fc=2./(24.*60)
    nblocks=16.
    
    wind,fwind=mlab.psd(dc,NFFT=int(len(dc)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(dc)/nblocks/2.))
    ax[0].loglog(fwind,wind,label='original')
    ax[0].set_xlabel('f [1/min]')
    ax[0].set_ylabel('$G_{xx}(f)$')
    ax[0].legend()
    ax[0].grid(True)

    t=np.linspace(0,len(dc)/24.,len(dc))
    ax[1].plot(t,dc,label='unfiltered')

    fNyq=1./2.
    fp = 0.042*24./freq
    fs = 0.42*24./freq
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,dbu, db)
    b, a = signal.ellip(n, dbu, db, fn, 'high')
    y = signal.filtfilt(b,a,dc)

    ax[1].plot(t,y,label='filtered')
    ax[1].set_ylabel('Temperature ($^\circ C$)')
    ax[1].set_xlabel('Days')
    ax[1].legend()
    ax[1].set_xlim(0,30/24.*freq)
    ax[1].grid(True)
    wind2,fwind2=mlab.psd(y,NFFT=int(len(y)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(y)/nblocks/2.))
    ax[0].loglog(fwind2,wind2,label='filtered')
    
    return ax,t,y
fig,ax=plt.subplots(1,2)
fig.set_size_inches(15,5)

ax[:],t,xHF1=highPass(temp[1,:]*1.,ax[:],24)
```

```python
tau = np.arange(0,24.,1.)
temp2=temp*1.-np.mean(temp,1)[:, np.newaxis]

def getLagCorr(s1,s2,tau):
    rho = 0.*tau
    for ind,t in enumerate(tau):
    
        if t>0:
            rho[ind]=np.mean(s1[t:]*s2[:-t])
            rho[ind]=rho[ind]/(np.std(s1[t:-t])*np.std(s2[t:-t]))
        else:
            rho[ind]=np.mean(s1*s2)
            rho[ind]=rho[ind]/(np.std(s1)*np.std(s2))
    return rho
```
