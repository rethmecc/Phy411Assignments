```python
def plotPDF(start=0,end=len(jbClean[:,1]),years=3,nbins=100):
    
    time=np.zeros(len(jbClean),dtype=bool)
    time[start:end]=True
    if years>1:
        time[start+60*24*365:end+60*24*365]=True
    if years>2:
        time[start+60*24*365*2:end+60*24*365*2]=True
    
    fig,ax=plt.subplots(1,1,figsize=(20,10))
    junk=ax.hist(jbClean[time,1],histtype='step',bins=nbins,label='James Bay',normed=True,color='b')
    junk=ax.hist(dcClean[time,1],histtype='step',bins=nbins,label='Deep Cove',normed=True,color='g')
    junk=ax.hist(jmClean[time,1],histtype='step',bins=nbins,label='John Muir',normed=True,color='r')
    junk=ax.hist(uiClean[time,1],histtype='step',bins=nbins,label='UVic ISC',normed=True,color='c')
    ax.legend()
    ax.set_ylim(0,0.15)
    ax.set_xlim([-10,40])
    ax.set_xlabel('Temperature ($^\circ$C)')

    fig2,bx=plt.subplots(2,2,figsize=(20,10))
    junk=bx[0][0].hist(jbClean[time,1],histtype='step',bins=nbins,label='James Bay',normed=True,color='b')
    junk=bx[0][1].hist(dcClean[time,1],histtype='step',bins=nbins,label='Deep Cove',normed=True,color='g')
    junk=bx[1][0].hist(jmClean[time,1],histtype='step',bins=nbins,label='John Muir',normed=True,color='r')
    junk=bx[1][1].hist(uiClean[time,1],histtype='step',bins=nbins,label='UVic ISC',normed=True,color='c')
    bx[0][0].legend()
    bx[0][0].set_ylim(0,0.15)
    bx[0][0].set_xlim([-10,40])
    bx[0][0].set_xlabel('Temperature ($^\circ$C)')
    bx[0][1].legend()
    bx[0][1].set_ylim(0,0.15)
    bx[0][1].set_xlim([-10,40])
    bx[0][1].set_xlabel('Temperature ($^\circ$C)')
    bx[1][0].legend()
    bx[1][0].set_ylim(0,0.15)
    bx[1][0].set_xlim([-10,40])
    bx[1][0].set_xlabel('Temperature ($^\circ$C)')
    bx[1][1].legend()
    bx[1][1].set_ylim(0,0.15)
    bx[1][1].set_xlim([-10,40])
    bx[1][1].set_xlabel('Temperature ($^\circ$C)')
    
    print 'James Bay mean = %1.2f +- %1.2f' %(np.mean(jbClean[time,1]),np.std(jbClean[time,1]))
    print 'Deep Cove mean = %1.2f +- %1.2f' %(np.mean(dcClean[time,1]),np.std(dcClean[time,1]))
    print 'John Muir mean = %1.2f +- %1.2f' %(np.mean(jmClean[time,1]),np.std(jmClean[time,1]))
    print 'UVic ISC mean = %1.2f +- %1.2f' %(np.mean(uiClean[time,1]),np.std(uiClean[time,1]))
    
    return fig,fig2
fig1,fig2=plotPDF()
fig1.savefig('PDFAllYear.pdf')
fig1.savefig('PDFAllYear2.pdf')
```

All PDFs look like they have 2 peaks. John Muir and Deep Cove appear the most different, as one would expect rom their locations. Deep Cove is at the northern end of Saanich Peninsula, while John Muir is in Sooke.
The mean temperature is more than a degree cooler in Sooke. Let's investigate the seasonality of this...

```python
fig1,fig2=plotPDF(60*24*171,60*24*265)
fig1.savefig('summer(171-265).pdf')
fig2.savefig('summer(171-265)2.pdf')
```

This is interesting. The Sooke PDF is much narrower than the others in the summer. The mean is also clearly much less: 2.5-3 degrees less. This station is unique in that it is on the west coast of the island. The ocean is most likely buffering the temperature - keeping it from getting too warm in the summer, and preventing temperature swings on a daily basis. The other stations do not have as much of a buffer, as it is more inland from the open ocean.

```python
fig1,fig2=plotPDF(0,60*24*90,3)
fig1.savefig('winter(0-90).pdf')
fig1.savefig('winter(0-90)2.pdf')
```

In the winter, the means are much more similar. Sooke and Deep Cove now have slightly wider distributions than the other stations. It looks like those stations get periods of near 0 temperatures that the other stations don't get.

```python
dc_=dc*1.
jb_=jb*1.
jm_=jm*1.
ui_=ui*1.

temps=np.zeros((len(dc),4))
temps[:,0]=dc[:,1]
temps[:,1]=jb[:,1]
temps[:,2]=jm[:,1]
temps[:,3]=ui[:,1]


T=len(dc[:,1])
for i in range(4):
    t=np.where(np.isfinite(temps[:,i]))[0]
    some=temps[t,i]
    f = interp1d(t, some)
    for j in range(T):
        if np.isfinite(temps[j,i])==False:
            vals=np.where(np.isfinite(temps[j,:]))[0]
            if np.shape(vals)[0]==0:
                temps[j,i]=f(i)
            else:
                temps[j,i]=np.mean(temps[j,vals])
                
dc_[:,1]=temps[:,0]
jb_[:,1]=temps[:,1]
jm_[:,1]=temps[:,2]
ui_[:,1]=temps[:,3]
```

```python
def fitQ3():
    fig,(ax,bx)=plt.subplots(1,2)
    fig.set_size_inches(15,5)

    ax.plot(dcT,jbT,'.')
    ax.plot(x,yFit,'-')
    ax.set_xlabel('Deep Cove Temp.');ax.set_ylabel('James Bay Temp.');ax.set_title('James Bay vs Deep Cove Weather Data')

    (n,xx,yy,im)=bx.hist2d(x,y,bins=100,cmap=cm.Reds,cmin=0.,normed=True)
    plt.colorbar(im,shrink=0.5)
    bx.set_xlabel('Deep Cove Temp.');bx.set_ylabel('James Bay Temp.');bx.set_title('James Bay vs Deep Cove Weather Data')
    bx.plot(x,yFit,'-')
    
    return ax,bx
```

```python
x=dcT*1.;y=jbT*1.
xy = x*y
b = (np.sum(x*y)-T*np.mean(x)*np.mean(y))/(np.sum(x**2)-T*np.mean(x)**2)
print 'Slope is %1.2f'%b
a = np.mean(y)-b*np.mean(x)
print 'Intercept is %1.2f'%a
yFit=b*x+a

syx = np.sqrt(np.sum((y-yFit)**2)/(T-2))

bdn,bup=stats.t.interval(0.95,df=T-2,scale=syx)
bb=np.array([bdn,bup])
bb=bb/np.sqrt(np.sum((x-np.mean(x))**2))
print 'Slope b is between %f and %f with 95%% probability'%(b+bb[0],b+bb[1])

ax,bx=fitQ3()
xx=np.sort(x)
yhathigh=(b+bb[1])*(xx-np.mean(x))+np.mean(y)
yhatlow=(b+bb[0])*(xx-np.mean(x))+np.mean(y)
ax.fill_between(xx,yhathigh,yhatlow,alpha=0.4)
```

```python
tau = np.arange(0,24.*60.,1.)
dc0=dcT*1.-np.nanmean(dcT)
jb0=jbT*1.-np.nanmean(jbT)
jm0=jmT*1.-np.nanmean(jmT)
ui0=uiT*1.-np.nanmean(uiT)

def getLagCorr(s1,s2,tau):
    rho = 0.*tau
    for ind,t in enumerate(tau):
    
        if t>0:
            rho[ind]=np.mean(s1[t:]*s2[:-t])
            rho[ind]=rho[ind]/(np.std(s1[t:-t])*np.std(s2[t:-t]))
        else:
            rho[ind]=np.mean(s1*s2)
            rho[ind]=rho[ind]/(np.std(s1)*np.std(s2))
    return rho

rho_dcjb = getLagCorr(dc0,jb0,tau)
rho_dcjm = getLagCorr(dc0,jm0,tau)
rho_dcui = getLagCorr(dc0,ui0,tau)
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,10))
ax.plot(tau/60.,rho_dcjb,label='DC & JB')
ax.plot(tau/60.,rho_dcjm,label='DC & JB')
ax.plot(tau/60.,rho_dcui,label='DC & JB')
ax.legend()
ax.set_xlabel(r'$\tau [hrs]$')
ax.set_ylabel(r'$\rho_{xy}(\tau)$')
max1=tau[np.where(rho_dcjb==min(rho_dcjb))[0]]/60.
max2=tau[np.where(rho_dcjm==min(rho_dcjm))[0]]/60.
max3=tau[np.where(rho_dcui==min(rho_dcui))[0]]/60.
print (max1-12)*60.,(max2-12)*60.,(max3-12)*60.
```

```python
def bandPass24(dc,ax):
    fc=2./(60.*24.)
    nblock=16.
    
    wind,fwind=mlab.psd(dc,NFFT=int(len(dc)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(dc)/nblocks/2.))
    ax[0].loglog(fwind,wind,label='unfiltered')
    ax[0].set_xlabel('f [1/min]')
    ax[0].set_ylabel('$G_{xx}(f)$')
    ax[0].grid(True)

    t=np.linspace(0,len(dc)/60./24.,len(dc))
    ax[1].plot(t,dc,label='unfiltered')
    ax[1].set_xlim(0,10)

    fNyq=1./2.
    fp = 0.0007
    fs = 0.004
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 100.)
    b,a=signal.ellip(n,.1,100.,fn)
    y = signal.filtfilt(b,a,dc)

    fp = 0.0007
    fs = 0.007
    n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 100.)
    b, a = signal.ellip(n, 0.1, 100., fn, 'high')
    y = signal.filtfilt(b,a,y)

    ax[1].plot(t,y,label='filtered')
    ax[1].set_ylabel('Temperature ($^\circ C$)')
    ax[1].set_xlabel('Days')
    ax[1].legend()
    ax[1].set_xlim(0,10)
    wind2,fwind2=mlab.psd(y,NFFT=int(len(y)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(y)/nblocks/2.))
    ax[0].loglog(fwind2,wind2,label='filtered')
    ax[0].legend()
    
    return ax,t,y

fig,bx=plt.subplots(1,2)
fig2,ax=plt.subplots(3,2)
fig.set_size_inches(15,5)
fig2.set_size_inches(15,15)

bx[:],t,dcF=bandPass24(dcT,bx[:])
ax[0,:],t,jbF=bandPass24(jbT,ax[0,:])
ax[1,:],t,jmF=bandPass24(jmT,ax[1,:])
ax[2,:],t,uiF=bandPass24(uiT,ax[2,:])

fig.savefig('filtered.pdf')
```

```python
def plotPDFFilt(start=0,end=len(jbF),years=3,nbins=100):
    ylims=(0,0.40)
    xlims=(-10,10)
    
    time=np.zeros(len(jbF),dtype=bool)
    time[start:end]=True
    if years>1:
        time[start+60*24*365:end+60*24*365]=True
    if years>2:
        time[start+60*24*365*2:end+60*24*365*2]=True
    
    fig,ax=plt.subplots(1,1,figsize=(20,10))
    junk=ax.hist(jbF[time],histtype='step',bins=nbins,label='James Bay',normed=True,color='b')
    junk=ax.hist(dcF[time],histtype='step',bins=nbins,label='Deep Cove',normed=True,color='g')
    junk=ax.hist(jmF[time],histtype='step',bins=nbins,label='John Muir',normed=True,color='r')
    junk=ax.hist(uiF[time],histtype='step',bins=nbins,label='UVic ISC',normed=True,color='c')
    ax.legend()
    ax.set_ylim(ylims)
    ax.set_xlim(xlims)
    ax.set_xlabel('Temperature ($^\circ$C)')

    fig2,bx=plt.subplots(2,2,figsize=(20,10))
    junk=bx[0][0].hist(jbF[time],histtype='step',bins=nbins,label='James Bay',normed=True,color='b')
    junk=bx[0][1].hist(dcF[time],histtype='step',bins=nbins,label='Deep Cove',normed=True,color='g')
    junk=bx[1][0].hist(jmF[time],histtype='step',bins=nbins,label='John Muir',normed=True,color='r')
    junk=bx[1][1].hist(uiF[time],histtype='step',bins=nbins,label='UVic ISC',normed=True,color='c')
    bx[0][0].legend()
    bx[0][0].set_ylim(ylims)
    bx[0][0].set_xlim(xlims)
    bx[0][0].set_xlabel('Temperature ($^\circ$C)')
    bx[0][1].legend()
    bx[0][1].set_ylim(ylims)
    bx[0][1].set_xlim(xlims)
    bx[0][1].set_xlabel('Temperature ($^\circ$C)')
    bx[1][0].legend()
    bx[1][0].set_ylim(ylims)
    bx[1][0].set_xlim(xlims)
    bx[1][0].set_xlabel('Temperature ($^\circ$C)')
    bx[1][1].legend()
    bx[1][1].set_ylim(ylims)
    bx[1][1].set_xlim(xlims)
    bx[1][1].set_xlabel('Temperature ($^\circ$C)')
    
    print 'James Bay mean = %1.2f +- %1.2f' %(np.mean(jbF[time]),np.std(jbF[time]))
    print 'Deep Cove mean = %1.2f +- %1.2f' %(np.mean(dcF[time]),np.std(dcF[time]))
    print 'John Muir mean = %1.2f +- %1.2f' %(np.mean(jmF[time]),np.std(jmF[time]))
    print 'UVic ISC mean = %1.2f +- %1.2f' %(np.mean(uiF[time]),np.std(uiF[time]))
    
    return fig,fig2
fig1,fig2=plotPDFFilt(60*24*171,60*24*265)
fig1.savefig('summerFiltered.pdf')
fig2.savefig('summerFiltered2.pdf')
```

Filtering the data for the 24 hour signal makes the buffering effect of the ocean even more apparent. There are two peaks in the plots above. They are the daily highs and lows. The difference between the highs and lows is smaller at Sooke than at the other stations. UVic has one of the highest differences, which makes sense since it would have the least exposure the the water.

```python
fig1,fig2=plotPDFFilt(60*24*0,60*24*90)
fig1.savefig('winterFiltered.pdf')
fig2.savefig('winterFiltered2.pdf')
```

The buffering is the most apparent in the winter. The fact that there is a very large peak at 0 in the filtered Sooke PDF implies that there are days when the daily signal is almost non-existant, i.e., the temperature does not follow much of a daily cycle. At UVic, this is definitely not the case. The two peaks clearly show a strong daily cycle. It is interesting to note that UVic has a wider PDF than Sooke. This may be due to changing wind direction between day and night on some days in Sooke, which again would be associated with the ocean.
There is a very narow peak at Deep Cove. Weather patterns at the top of the Peninsula must somehow hold the temperature very constant on some days. Or this could just be an issue of low statistics affecting the filtering.

```python
def spec(x):
    #T=np.shape(hour[0,5:])[0]
    T=np.shape(dcT)[0]
    t=np.arange(T)
    N = np.shape(x)[0]
    n = int(np.floor(N/20))
    nfft=256.*8.*64
    Gxx = np.zeros((20*10,nfft/2+1))
    num=0
    tbin = np.zeros(20*10)
    for ind in range(0,N-n,int(np.floor(n/10))):
    #    print ind
        px,f=np.squeeze(mlab.psd(x[ind+np.arange(0,n)],nfft,Fs=1/np.median(np.diff(t))))
        Gxx[num,:]=px
        num=num+1
        tbin[num]=np.mean(t[ind+np.arange(0,n)])
    tbin=tbin[0:num]
    Gxx=Gxx[0:num,:]
    fig,ax=plt.subplots(1,1)
    ax.loglog(f,Gxx.transpose())
    plt.xlabel('f [Hz]');
    plt.ylabel(r'$G_{xx} \mathrm{[m^2 Hz^{-1}]}$')
    
    return tbin,f,Gxx

def cont(tbin,f,Gxx):
    fig,ax=plt.subplots(1,1)
    plt.axes(ax)
    print f
    pcm=ax.pcolormesh(tbin,f,np.log10(Gxx.transpose()))
    plt.xlim((0,25000))
    #plt.ylim((0,0.5))
    pcm.set_clim((-2,1.5))
    plt.colorbar(pcm)
    plt.title(r'$log_{10}(G_{xx})$')
    plt.xlabel('t [s]')
    plt.ylabel('F [Hz]')
```

```python
tbin,f,Gxx=spec(dcT[:60*24*30*12]*1.)
```

```python
cont(tbin,f,Gxx)
```
