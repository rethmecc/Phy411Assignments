```python
#Load libraries
import scipy, scipy.stats
import pylab
%matplotlib inline
import matplotlib.pyplot as plt7
import matplotlib.mlab as mlab
```

```python
#Set probabilities
p=0.5
q=1-p

#create figure
fig,ax=plt.subplots(1,3,figsize=(20,6))

#Create array of points representing steps to the right.
x5 = scipy.linspace(0,5,6)
x10 = scipy.linspace(0,10,11)
x20= scipy.linspace(0,20,21)

#Calculate binomial probabilties at each point in array x, with N=5,10,20 and probability p.
pmf5 = scipy.stats.binom.pmf(x5,5,p)
pmf10 = scipy.stats.binom.pmf(x10,10,p)
pmf20 = scipy.stats.binom.pmf(x20,20,p)

#Plot results. Since we are plotting m, not n1, we need to use m=2n1-N to convert the x-axis to m.
ax[0].plot((x5-2.5)*2,pmf5,'*')
ax[1].plot((x10-5)*2,pmf10,'*')
ax[2].plot((x20-10)*2,pmf20,'*')

#Set the mean for the Gaussian distributions.
mean=p-q

#Calculate sigma, generate x-axis, and plot Gaussian. The Gaussian is multiplied by 2 to
#compensate for the fact that we are plotting m instead of n, which changes dx to 2 steps instead of 1.
#This block is for N=5
sigma = scipy.sqrt(4*p*q*5)
x=scipy.linspace(-5,5,11)
ax[0].plot(x,2*mlab.normpdf(x,mean,sigma))

#Again for N=10
sigma = scipy.sqrt(4*p*q*10)
x=scipy.linspace(-10,10,21)
ax[1].plot(x,2*mlab.normpdf(x,mean,sigma))

#N=20
sigma = scipy.sqrt(4*p*q*20)
x=scipy.linspace(-20,20,41)
ax[2].plot(x,2*mlab.normpdf(x,mean,sigma))

#Set titles
ax[0].set_title('N=5')
ax[1].set_title('N=10')
ax[2].set_title('N=20')
fig.suptitle('p=0.5',fontsize=20)
```

```python
#Set probabilities
p=2./3.
q=1-p

#Create figure
fig,ax=plt.subplots(1,3,figsize=(20,6))

#Create array of points for x-axis
x5 = scipy.linspace(0,5,6)
x10 = scipy.linspace(0,10,11)
x20= scipy.linspace(0,20,21)

#Generate binomials
pmf5 = scipy.stats.binom.pmf(x5,5,p)
pmf10 = scipy.stats.binom.pmf(x10,10,p)
pmf20 = scipy.stats.binom.pmf(x20,20,p)

#Plot binomials
ax[0].plot((x5-2.5)*2,pmf5,'*')
ax[1].plot((x10-5)*2,pmf10,'*')
ax[2].plot((x20-10)*2,pmf20,'*')

#Calculate mean and sigma for Gaussian and plot it for N=5.
mean=5*(p-q)
sigma = scipy.sqrt(4*p*q*5)
x=scipy.linspace(-5,5,11)
ax[0].plot(x,2*mlab.normpdf(x,mean,sigma))

#Gaussian for N=10
mean=10*(p-q)
sigma = scipy.sqrt(4*p*q*10)
x=scipy.linspace(-10,10,21)
ax[1].plot(x,2*mlab.normpdf(x,mean,sigma))

#N=20
mean=20*(p-q)
sigma = scipy.sqrt(4*p*q*20)
x=scipy.linspace(-20,20,41)
ax[2].plot(x,2*mlab.normpdf(x,mean,sigma))

#Titles
ax[0].set_title('N=5')
ax[1].set_title('N=10')
ax[2].set_title('N=20')
fig.suptitle('p=2/3',fontsize=20)
```
