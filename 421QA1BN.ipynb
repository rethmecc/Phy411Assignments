```python
import scipy, scipy.stats
import pylab
%matplotlib inline
import matplotlib.pyplot as plt7
import matplotlib.mlab as mlab
```

```python
p=0.5
q=1-p

fig,ax=plt.subplots(1,3,figsize=(20,6))
x5 = scipy.linspace(0,5,6)
x10 = scipy.linspace(0,10,11)
x20= scipy.linspace(0,20,21)
pmf5 = scipy.stats.binom.pmf(x5,5,p)
pmf10 = scipy.stats.binom.pmf(x10,10,p)
pmf20 = scipy.stats.binom.pmf(x20,20,p)
ax[0].plot((x5-2.5)*2,pmf5,'.')
ax[1].plot((x10-5)*2,pmf10,'.')
ax[2].plot((x20-10)*2,pmf20,'.')

mean=p-q
sigma = scipy.sqrt(4*p*q*5)
x=scipy.linspace(-5,5,11)
ax[0].plot(x,2*mlab.normpdf(x,mean,sigma))

sigma = scipy.sqrt(4*p*q*10)
x=scipy.linspace(-10,10,21)
ax[1].plot(x,2*mlab.normpdf(x,mean,sigma))

sigma = scipy.sqrt(4*p*q*20)
x=scipy.linspace(-20,20,41)
ax[2].plot(x,2*mlab.normpdf(x,mean,sigma))

#mlab.normpdf()
```
