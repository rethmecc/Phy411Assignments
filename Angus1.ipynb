```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import matplotlib.mlab as mlab
N=1000
dt = 0.1
t = np.arange(0,N*dt,dt)
x0 = np.random.randn(N)
#Bandwidth = 0.5 Hz
B=0.5
# filter to make Bandlimited signal
fp=B*0.9
fs = B
fNyq=1/2./dt
n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 60.)
b,a=signal.ellip(n,.1,60.,fn)
xBL = signal.filtfilt(b,a,x0)
fig,ax=plt.subplots(2,1,figsize=(15,8))

ax[0].plot(t,x0,color='0.5',label='full-spectrum signal')
ax[0].plot(t,xBL,color='k',label='band-limited signal')
#ax[0].set_xlim([0,30])
ax[0].set_xlabel('time [s]');ax[0].set_ylabel('x [V]')
ax[0].legend(fontsize='small',loc=0)

nfft=512
pxx,f=mlab.psd(x0,NFFT=nfft,Fs=1,noverlap=nfft/2.,window=mlab.window_hanning)
pxBL,f=mlab.psd(xBL,NFFT=nfft,Fs=1,noverlap=nfft/2.,window=mlab.window_hanning)
print pxBL
ax[1].loglog(f,pxx,color='0.5')
ax[1].loglog(f,pxBL,color='k')
plt.axvline(x=B,linestyle='--',color='k');ax[1].text(B,0.05,' B=0.5 Hz')
ax[1].set_xlabel('f [Hz]');ax[1].set_ylabel('$G_{xx} [V^2 Hz^{-1}]$')
plt.tight_layout()
```
