## Q2: Demonstrate the central limit theorem for a very unfair coin flip.

Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=-1/2$ and heads $x=+1/2$.  

**1** What are the mean, $\mu_x$, and variance $\sigma_x^2$ of this process?

<hr>

Probability of heads = 2/3; prob. of tails = 1/3

Mean = 1/6

Variance = 2/9

```python
import math
Mean = 1./3.*(-0.5)+2./3.*0.5
Var = 1./3.*math.pow((-0.5-1./6.),2) + 2./3.*math.pow((0.5-1./6.),2)
print Mean
print Var
```

**2** Plot or represent the *probability distribution*  (or cumulative distrubution) of this process.
<hr>

```python
x = np.arange(-1.,1.,0.01)
P = x*0.
for i in range(0,2,1):
    P[x>(i-0.5)]=1./3.+2./3.*i
fig=plt.figure();ax=fig.add_subplot(1,1,1)
ax.plot(x,P)
ax.set_xlabel('x - coin value');ax.set_ylabel('P(x)');ax.set_ylim([-0.1,1.1])
```

**3** Write a function using the `np.random.norm` function to simulate the coin flip.  It shoudl return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
def Nflip(N,toprint=False,seed=None):
    r = np.random.rand(N)
    x = np.empty(N,float)
    heads=0.
    tails=0.
    for a in range (0,N):
        if (r[a]>2./3.):
            x[a]=-0.5
            tails=tails+1
        else:
            x[a]=0.5
            heads=heads+1
    if toprint:
        print heads/(tails+heads)
        print (heads-tails)/(tails+heads)
    return x

x=Nflip(10000,True,0)
fig=plt.figure(figsize=(12,4))
ax=fig.add_subplot(1,1,1)
ax.hist(x)
print np.mean(x)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{flip}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll discussed in the Lecture notes.
<hr>

```python
Nexpt=2000
x2mean = np.empty(Nexpt,float)
x16mean = np.empty(Nexpt,float)
x64mean = np.empty(Nexpt,float)
x256mean = np.empty(Nexpt,float)
for i in range (0,Nexpt):
    x2mean[i]=np.mean(Nflip(2))
    x16mean[i]=np.mean(Nflip(16))
    x64mean[i]=np.mean(Nflip(64))
    x256mean[i]=np.mean(Nflip(256))

bins=10
    
fig2=plt.hist(x2mean,bins,label='2 flips',histtype='step',normed=True)
fig16=plt.hist(x16mean,bins,label='16 flips',histtype='step',normed=True)
fig64=plt.hist(x64mean,bins,label='64 flips',histtype='step',normed=True)
fig256=plt.hist(x256mean,bins,label='256 flips',histtype='step',normed=True)
plt.legend(loc='upper left')

print np.mean(x2mean)
print np.mean(x16mean)
print np.mean(x64mean)
print np.mean(x256mean)
```

It can very easily be seen in the above histogram that as N->infinity, the distribution of the means approaches a gaussian centred at 1/6.

Like the dice roll case, the distribution of the means becomes narrower as N increases. The difference here is that the distribution also becomes more symmetric as N increases, while in the dice roll case the distribution was already symetric. This difference is due to the fact that the dice roll had a uniform distribution, while the unair coin toss does not.
Both this case and the dice roll case become more norrower and gaussian-like as N increases.

```python
import scipy.stats as stats
import matplotlib.mlab as mlab

fig2=plt.hist((x2mean-np.mean(x2mean))/np.std(x2mean),bins,label='2 flips',histtype='step',normed=True)
fig16=plt.hist((x16mean-np.mean(x16mean))/np.std(x16mean),bins,label='16 flips',histtype='step',normed=True)
fig64=plt.hist((x64mean-np.mean(x64mean))/np.std(x64mean),bins,label='64 flips',histtype='step',normed=True)
fig256=plt.hist((x256mean-np.mean(x256mean))/np.std(x256mean),bins,label='256 flips',histtype='step',normed=True)
plt.legend(loc='upper left')

x=np.linspace(-4,4,50)
y = stats.norm.cdf(x,0,1)
fig,ax=plt.subplots(1,1)
ax.hist((x2mean-np.mean(x2mean))/np.std(x2mean),bins,label='2 flips',histtype='step',normed=True,cumulative=True,align='right')
ax.hist((x16mean-np.mean(x16mean))/np.std(x16mean),bins,label='16 flips',histtype='step',normed=True,cumulative=True,align='right')
ax.hist((x64mean-np.mean(x64mean))/np.std(x64mean),bins,label='64 flips',histtype='step',normed=True,cumulative=True,align='right')
ax.hist((x256mean-np.mean(x256mean))/np.std(x256mean),bins,label='256 flips',histtype='step',normed=True,cumulative=True,align='right')
ax.plot(x,y,color='black',label='Normal Dist')
ax.legend(loc='upper left')
```
