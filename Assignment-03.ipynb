## Q1: Lag correlations of weather time series

````python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.stats as stats
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]
dcClean=dc[np.isfinite(dc)]
````

**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xy}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
dc0=dc-np.mean(dcClean)
dc0Clean=dc0[np.isfinite(dc0)]
lag=np.arange(1,30*24+1)
cxx=lag*0.
for i,tau in enumerate(lag):
    temp=dc0[:-tau]*dc0[tau:]
    temp=temp[np.isfinite(temp)]
    cxx[i]=np.mean(temp)
pxx=cxx/np.var(dc0Clean)
fig,ax=plt.subplots(1,1)
fig.set_size_inches(15,5)
ax.plot(lag/24.,pxx)
ax.set_xlabel(r'$\tau [day]$')
ax.set_ylabel(r'$\rho_{xx}$')
ax.xaxis.set_ticks(np.arange(0., 30., 1.))
ax.xaxis.grid(True)
````

There is a strong correlation for lags that are whole numbers of days. This makes sense, since temperatures will follow the same general paturn from day to day. It can also be seen that this correlation shrinks as time goes on, probably because of longer term temperature variations within a given 30 day period.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
summer=dc[(31+28+31+30+31+21)*24:(31+28+31+30+31+30+31+31+21)*24]
summerClean=summer[np.isfinite(summer)]
summer0=summer-np.mean(summerClean)
summer0Clean=summer0[np.isfinite(summer0)]
winter=dc[(31+28+31+30+31+30+31+31+30+31+30+21)*24:(31+28+31+30+31+30+31+31+30+31+30+31+31+28+21)*24]
winterClean=winter[np.isfinite(winter)]
winter0=winter-np.mean(winterClean)
winter0Clean=winter0[np.isfinite(winter0)]
lag=np.arange(1,30*24+1)
cxxS=lag*0.
cxxW=lag*0.
for i,tau in enumerate(lag):
    temp=summer0[:-tau]*summer0[tau:]
    temp=temp[np.isfinite(temp)]
    cxxS[i]=np.mean(temp)
for i,tau in enumerate(lag):
    temp=winter0[:-tau]*winter0[tau:]
    temp=temp[np.isfinite(temp)]
    cxxW[i]=np.mean(temp)
pxxS=cxxS/np.var(summer0Clean)
pxxW=cxxW/np.var(winter0Clean)
fig,ax=plt.subplots(1,1)
fig.set_size_inches(20,5)
ax.plot(lag/24.,pxxS,label='Summer')
ax.plot(lag/24.,pxxW,label='Winter')
ax.legend()
ax.set_xlabel(r'$\tau [day]$')
ax.set_ylabel(r'$\rho_{xx}$')
ax.xaxis.set_ticks(np.arange(0., 30., 1.))
ax.xaxis.grid(True)
````

The difference between the highs and lows in the summer seem to be greater than that in the winter.
