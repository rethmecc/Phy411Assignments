## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#1.-----------------------------------------
N=2048.
dt=1.
x=random.randn(N)
f = np.arange(N/2)/N
X = dt*np.fft.fft(x)
G = (2./N)*np.real(np.conj(X[:N/2])*X[:N/2])

end=np.array([0,60])
nbins=60
points=np.arange(0,end[1])
chi=stats.chi2.pdf(points,2)

fig,ax=plt.subplots(1,1)
fig.set_size_inches(20,10)
hand=ax.hist(G,bins=nbins,normed=True,label='Raw',range=end,histtype='step')
ax.plot(points,chi,label='chi2 with df=2',color='b')

#2.-----------------------------------------
nblock=4.

GCan=np.zeros((nblock,N/nblock/2.+1))
if nblock<2.:
    GCan[1,:]=mlab.psd(x, NFFT=N/nblock, Fs=1)[0][:,0]
else:
    for ind in range(int(nblock)):
        GCan[ind,:]=mlab.psd(x, NFFT=N/nblock, Fs=1)[0]
print np.shape(GCan)

fN = 0.5
Gtheory = 1./fN
Gblockm=np.mean(GCan,axis=0)
hand=ax.hist(Gblockm*nblock*2/Gtheory,bins=nbins,normed=True,label='Block average with no overlap (nblocks=%2.0f)'%nblock,range=end,histtype='step',color='r')

chi=stats.chi2.pdf(points,nblock*2)

hand=ax.plot(points,chi,label='chi2 with df=%2.0f'%(nblock*2),color='r')

#3.-----------------------------------------
Gcanned,fcanned = mlab.psd(x,NFFT=N/nblock,Fs=1.,window=mlab.window_hanning,noverlap=N/nblock/2.)
hand=ax.hist(Gcanned*nblock*2/Gtheory,bins=nbins,normed=True,label='Block average with 50 percent overlap (nblocks=%2.0f)'%nblock,range=end,histtype='step',color='c')
hand=ax.plot(points,chi,'--',label='chi2 with df=%2.0f'%(nblock*2),color='c')

#Now with higher nblocks
#2.-----------------------------------------
nblock=16.

GCan=np.zeros((nblock,N/nblock/2.+1))
if nblock<2.:
    GCan[1,:]=mlab.psd(x, NFFT=N/nblock, Fs=1)[0][:,0]
else:
    for ind in range(int(nblock)):
        GCan[ind,:]=mlab.psd(x, NFFT=N/nblock, Fs=1)[0]
print np.shape(GCan)

fN = 0.5
Gtheory = 1./fN
Gblockm=np.mean(GCan,axis=0)
hand=ax.hist(Gblockm*nblock*2/Gtheory,bins=nbins,normed=True,label='Block average with no overlap (nblocks=%2.0f)'%nblock,range=end,histtype='step',color='y')

chi=stats.chi2.pdf(points,nblock*2)

hand=ax.plot(points,chi,label='chi2 with df=%2.0f'%(nblock*2),color='g')

#3.-----------------------------------------
Gcanned,fcanned = mlab.psd(x,NFFT=N/nblock,Fs=1.,window=mlab.window_hanning,noverlap=N/nblock/2.)
hand=ax.hist(Gcanned*nblock*2/Gtheory,bins=nbins,normed=True,label='Block average with 50 percent overlap (nblocks=%2.0f)'%nblock,color='g',range=end,histtype='step')
hand=ax.plot(points,chi,'--',label='chi2 with df=%2.0f'%(nblock*2),color='y')
ax.legend()
ax.set_xlabel('$G_{xx}(f)$')
ax.set_ylabel('P')
```

The distribution is wider with a larger Nblock. The spectral estimates do indeed follow the chi2 distribution with 2Nblock degrees of freedom.

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
nblock=8.
fig,ax=plt.subplots(1,1)
fig.set_size_inches(15,5)

Gcanned,fcanned = mlab.psd(x,NFFT=N/nblock,Fs=1.,window=mlab.window_hanning,noverlap=N/nblock/2.)
hand=ax.hist(Gcanned*nblock*2/Gtheory,bins=nbins,normed=True,label='Block average with 50 percent overlap (nblocks=%2.0f)'%nblock,color='g',range=end,histtype='step')

chi2=stats.chi2.pdf(points,nblock*2)
chi1811=stats.chi2.pdf(points,nblock*18/11.)
hand=ax.plot(points,chi2,label='chi2 with df=2NBlock')
hand=ax.plot(points,chi1811,label='chi2 with df=18NBlock/11')
ax.legend()
ax.set_xlabel('$G_{xx}(f)$')
ax.set_ylabel('P')
```

Strangely, 2NBlocks fits much better than the 'correct' theoretical value.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
```

```python
print len(dc)
nblocks=16.
GCan=np.zeros((nblock,len(dc)/nblocks/2.+1))
fCan=np.zeros((nblock,len(dc)/nblocks/2.+1))

for ind in range(int(nblock)):
    GCan[ind,:],fCan=mlab.psd(dc, NFFT=int(len(dc)/nblocks), Fs=1)
GCanm=np.mean(GCan,axis=0)
fig,ax=plt.subplots(2,1)
fig.set_size_inches(15,10)
ax[1].loglog(fCan,GCanm,label='Block averaged')


Gcanned,fcanned = mlab.psd(dc,NFFT=len(dc),Fs=1.)
ax[0].loglog(fcanned,Gcanned)
ax[0].set_title('Raw')
ax[0].set_xlabel('f [1/hr]')
ax[0].set_ylabel('$G_{xx}(f)$')

wind,fwind=mlab.psd(dc,NFFT=int(len(dc)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(dc)/nblocks/2.))
ax[1].loglog(fwind,wind,label='Hanning windowed')
ax[1].set_title('Windowed')
ax[1].set_xlabel('f [1/hr]')
ax[1].set_ylabel('$G_{xx}(f)$')
ax[1].legend()
```

There is slightly less leakage in the Hanning windowed spectrum than in the block averaged spectrum.
