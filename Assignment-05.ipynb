## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
reps=100
N=2048.
fN = 0.5
Gtheory = 1./fN
nblock=16.
nblock2=4.
end=np.array([0,60])
points=np.linspace(0,4,100)
nbins=60

G1=np.zeros((N/2.,reps))
G2=np.zeros((N/2./nblock,reps))
G3=np.zeros((N/2./nblock,reps))
G4=np.zeros((N/2./nblock2,reps))
G5=np.zeros((N/2./nblock2,reps))
for i in range(reps):
    x=random.randn(N)
    g1,f1 = mlab.psd(x,NFFT=N,Fs=1.,window=mlab.window_none)
    g2,f2 = mlab.psd(x, NFFT=N/nblock, Fs=1,window=mlab.window_none)
    g3,f3 = mlab.psd(x,NFFT=N/nblock,Fs=1.,window=mlab.window_hanning,noverlap=N/nblock/2.)
    g4,f4 = mlab.psd(x, NFFT=N/nblock2, Fs=1,window=mlab.window_none)
    g5,f5 = mlab.psd(x,NFFT=N/nblock2,Fs=1.,window=mlab.window_hanning,noverlap=N/nblock2/2.)
    G1[:,i]=g1[1:].flatten()/Gtheory
    G2[:,i]=g2[1:].flatten()/Gtheory
    G3[:,i]=g3[1:].flatten()/Gtheory
    G4[:,i]=g4[1:].flatten()/Gtheory
    G5[:,i]=g5[1:].flatten()/Gtheory
    

G1=G1.flatten()
G2=G2.flatten()
G3=G3.flatten()
G4=G4.flatten()
G5=G5.flatten()

fig,ax=plt.subplots(1,1)
fig.set_size_inches(20,10)

hand=ax.hist(G1,normed=True,bins=nbins,histtype='step',range=[0,4],color='b',label='Raw power spectrum')
hand=ax.hist(G2,normed=True,bins=nbins,histtype='step',range=[0,4],color='g',label='Block averaged with %2.0f blocks'%nblock)
hand=ax.hist(G3,normed=True,bins=nbins,histtype='step',range=[0,4],color='r',label='Hanning window with %2.0f blocks'%(nblock*2-1))
hand=ax.hist(G4,normed=True,bins=nbins,histtype='step',range=[0,4],color='c',label='Block averaged with %2.0f blocks'%nblock2)
hand=ax.hist(G5,normed=True,bins=nbins,histtype='step',range=[0,4],color='m',label='Hanning window with %2.0f blocks'%(nblock2*2-1))

chi1=stats.chi2.pdf(points,2,scale=1./2.)
chi2=stats.chi2.pdf(points,nblock*2,scale=1./(2*nblock))
chi3=stats.chi2.pdf(points,(nblock*2-1)*2,scale=1./((nblock*2-1)*2))
chi4=stats.chi2.pdf(points,nblock2*2,scale=1./(2*nblock2))
chi5=stats.chi2.pdf(points,(nblock2*2-1)*2,scale=1./((nblock2*2-1)*2))

hand=ax.plot(points,chi1,color='b',label='$\chi^2_2$')
hand=ax.plot(points,chi2,color='g',label='$\chi^2_{%2.0f}$'%(2*nblock))
hand=ax.plot(points,chi3,color='r',label='$\chi^2_{%2.0f}$'%(2*(2*nblock-1)))
hand=ax.plot(points,chi4,color='c',label='$\chi^2_{%2.0f}$'%(2*nblock2))
hand=ax.plot(points,chi5,color='m',label='$\chi^2_{%2.0f}$'%(2*(2*nblock2-1)))

ax.legend()
ax.set_xlabel('$G_{xx}(f)$')
ax.set_ylabel('P')
```

The distribution is narrower with a larger Nblock. The spectral estimates do indeed follow the chi2 distribution with 2Nblock degrees of freedom.

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
chi6=stats.chi2.pdf(points,(nblock*2-1)*18./11.,scale=1./((nblock*2-1)*18./11.))

fig,ax=plt.subplots(1,1)
fig.set_size_inches(20,10)

hand=ax.hist(G3,normed=True,bins=nbins,histtype='step',range=[0,4],color='r',label='Hanning window with %2.0f blocks'%(nblock*2-1))
hand=ax.plot(points,chi3,color='g',label='$\chi^2_{%2.0f}$'%(2*(2*nblock-1)))
hand=ax.plot(points,chi6,color='r',label='$\chi^2_{%2.4f}$'%(18./11.*(2*nblock-1)))

ax.legend()
ax.set_xlabel('$G_{xx}(f)$')
ax.set_ylabel('P')
```

The corect theoretical value of 18NBlocks/11 fits better than 2NBlocks.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
```

```python
print len(dc)
nblocks=16.
GCan=np.zeros((nblock,len(dc)/nblocks/2.+1))
fCan=np.zeros((nblock,len(dc)/nblocks/2.+1))

for ind in range(int(nblock)):
    GCan[ind,:],fCan=mlab.psd(dc, NFFT=int(len(dc)/nblocks), Fs=1,window=mlab.window_none)
GCanm=np.mean(GCan,axis=0)
fig,ax=plt.subplots(2,1)
fig.set_size_inches(15,10)
ax[1].loglog(fCan,GCanm,label='Block averaged')


Gcanned,fcanned = mlab.psd(dc,NFFT=len(dc),Fs=1.,window=mlab.window_none)
ax[0].loglog(fcanned,Gcanned)
ax[0].set_title('Raw')
ax[0].set_xlabel('f [1/hr]')
ax[0].set_ylabel('$G_{xx}(f)$')

wind,fwind=mlab.psd(dc,NFFT=int(len(dc)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(dc)/nblocks/2.))
ax[1].loglog(fwind,wind,label='Hanning windowed with $50\%$ overlap')
ax[1].set_title('Windowed')
ax[1].set_xlabel('f [1/hr]')
ax[1].set_ylabel('$G_{xx}(f)$')
ax[1].legend()
```

There is slightly less leakage in the Hanning windowed spectrum than in the block averaged spectrum.

<div style='background:#F0F0F0'> **2** Now do the same thing to the intergal of the Deep Cove data with time: $$y(t)=\int_o^t x(t')\ \mathrm{d}t'$$  This is obviously a silly thing to do, but compare the spectra and comment on the difference between the Hanning window and the non-Hanning windowed data.  </div>

```python
y=np.zeros(len(dc)+1)
dt=1
for i in range(len(dc)+1):
    y[i]=np.sum(dc[:i]*dt)

fig,ax=plt.subplots(2,1)
fig.set_size_inches(15,10)

GCan=np.zeros((nblock,len(dc)/nblocks/2.+1))
fCan=np.zeros((nblock,len(dc)/nblocks/2.+1))

for ind in range(int(nblock)):
    GCan[ind,:],fCan=mlab.psd(y, NFFT=int(len(y)/nblocks), Fs=1,window=mlab.window_none)
GCanm=np.mean(GCan,axis=0)
ax[1].loglog(fCan,GCanm,label='Block averaged')

Gcanned,fcanned = mlab.psd(y,NFFT=len(y),Fs=1.)
ax[0].loglog(fcanned,Gcanned,label='Raw psd')
ax[0].set_title('Raw')
ax[0].set_xlabel('f [1/hr]')
ax[0].set_ylabel('$G_{yy}(f)$')

wind,fwind=mlab.psd(y,NFFT=int(len(y)/nblocks),Fs=1.,window=mlab.window_hanning,noverlap=int(len(y)/nblocks/2.))
ax[1].loglog(fwind,wind,label='Hanning windowed with $50\%$ overlap')
ax[1].set_title('Windowed')
ax[1].set_xlabel('f [1/hr]')
ax[1].set_ylabel('$G_{yy}(f)$')
ax[1].legend()
```

The windowed spectra absolutley require the overlappig Hanning window in this case. Block averaging does not yield any peaks.
