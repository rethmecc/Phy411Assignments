### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

```python
N=204800
dt=1.
b=25
x=np.random.randn(N)
n=np.random.randn(N)
y=np.zeros(N)
tAxis=np.arange(0,N,1)
for t in range(N):
    y[t]=0.05*n[t]+1./b*np.sum(x[(t-b):t])

fig,ax=plt.subplots(2,1)
fig.set_size_inches(15,15)
ax[0].plot(tAxis,x,label='input')
ax[0].plot(tAxis,y,label='output')
ax[1].plot(tAxis[500:1000],x[500:1000],label='input')
ax[1].plot(tAxis[500:1000],y[500:1000],label='output')
ax[0].legend()
ax[1].legend()
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel and the coherence squared in the middle panel and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

```python
def above(nblocks,uncert=False):
    Gxy,f=mlab.csd(x,y,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
    Gxx,f=mlab.psd(x,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
    Gnn,f=mlab.psd(n,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
    Gyy,f=mlab.psd(y,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)

    fig,ax=plt.subplots(3,1,figsize=(9,9))
    ax[0].loglog(f,Gxx,label='input')
    ax[0].loglog(f,Gnn,label='noise')
    ax[0].loglog(f,Gyy,label='output')
    ax[0].set_ylabel(r'$G(f)$')
    ax[0].legend(loc=0,fontsize='small')
    gamma2=abs(Gxy)**2/Gxx/Gyy
    ax[1].semilogx(f,gamma2,'k')
    ax[1].set_ylabel(r'$\gamma^2$')
    ax[1].set_ylim(-0.1,1.1)
    if uncert:
        ax[1].set_ylim(-1,3)
        uncertainty=np.sqrt(2)*(1+gamma2)/(np.abs(np.sqrt(gamma2))*np.sqrt(nblocks))
        ax[1].fill_between(f,gamma2+uncertainty,gamma2-uncertainty)
        ax[1].semilogx(f,gamma2-uncertainty)
        ax[1].semilogx(f,gamma2+uncertainty)

    ang = np.angle(Gxy)
    ax[2].semilogx(f,ang/np.pi,'k')
    ax[2].set_ylabel('$Phase / \pi$')
    ax[2].set_xlabel('$f / f_N$')
    ax[2].set_ylim(-1.1,1.1)
    
    return ax

above(10)
```

The lowest frequency peak is at 0.04Hz, due to the fact that b is 25s. There are peaks every n/25Hz where n is an integer from 1 to 25. This is due to the fact that the running average sees the input from up to 25 seconds ago, so every point is correlated with each of the previous 25 points.

They are coherant at low frequencies (below 0.04Hz) because below this value there is no contribution from the moving average. Near the n/25Hz frequencies the signals are incoherant because the output is referencing the previous values of the input, all of which are independant of each other.

The phase is 0 when the signals are coherant, and jumps between -$\pi$ and $\pi$ when they are incoherant.

$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.

## Q2: Shear turbulence data

```python
time=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/time.txt')
ac=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/ac.txt')
shear=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/s.txt')
```


<div style='background:#F0F0F0'>Download the data abaove. `time` is a measurement time in seconds. `shear` is a little less than 300 s of data from a turbulence shear probe. Units are $s^{-1}$. `ac` is a suitably scaled acceleration of the vehicle containing the shear probes.

**1** Plot the shear time series and the accelerometer time series.  Comment on any similarities and differences.  Comment on the stationarity of the time series.

```python
fig,ax=plt.subplots(1,1)
fig.set_size_inches(15,7)

ax.plot(time,shear,label='Shear')
ax.plot(time,ac,label='Accelerometer')
ax.legend()
ax.set_xlabel('time (s)')
```

Towards the end of the measurement (after about 200s), both times series get more noisy, which makes sense since turbulance and acceleration should be somewhat correlated in such an environment.

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence.  

```python
nblocks=50
N=len(time)
T=time[N-1]
dt=T/N
Gsa,f=mlab.csd(shear,ac,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
Gss,f=mlab.psd(shear,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
Gaa,f=mlab.psd(ac,NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)

gamma2=abs(Gsa)**2/Gss/Gaa

fig,ax=plt.subplots(2,1)
fig.set_size_inches(9,6)

ax[0].loglog(f,Gaa,label='Accelerometer')
ax[0].loglog(f,Gss,label='Shear')
ax[0].legend(loc='lower left')
ax[0].set_ylabel('G(f)')
ax[1].semilogx(f,gamma2)
ax[1].set_ylabel('$\gamma^2$')
ax[1].set_xlabel('f [Hz]')
```

There are 2 large peaks in common between the accelerometer spectrum and the shear spectrum. This is indicative of vibrations of the vehicle causing the shear to measure artificial "turbulance". Those same two peaks also appear quite prominently in the $\gamma^2$ plot, with $\gamma^2 = 1$ at 50Hz, indicating a total correlation between the measured turbulance and the vibration of the vehicle.

<div style='background:#F0F0F0'> **3** Compute a spectra that has the contamination from the vibration removed using the linear technique described in the notes. Comment on how well your correction worked.

```python
Grr=Gss-abs(Gsa)**2/Gaa

fig,ax=plt.subplots(1,1)
fig.set_size_inches(9,3)
ax.loglog(f,Gaa,label='Accelerometer')
ax.loglog(f,Gss,label='Shear')
ax.loglog(f,Grr,label='Vibration removed')
ax.legend(loc='lower left')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('G(f)')
```

The correction worked pretty well. There is till the peak at 50Hz, but it was reduced by a couple orders of magnitude.

<div style='background:#F0F0F0'>**4** Chose two separate periods of the time series by eye, one that is "turbulent" and one that is "quiet", and plot the raw and corrected shear spectra.  Comment on the differences and the effect of the correction on each spectral estimate.    

```python
nblocks=50
N=len(time)
T=time[N-1]
dt=T/N
TGsa,f=mlab.csd(shear[30000:],ac[30000:],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
TGss,f=mlab.psd(shear[30000:],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
TGaa,f=mlab.psd(ac[30000:],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
TGrr=TGss-abs(TGsa)**2/TGaa

QGsa,f=mlab.csd(shear[5500:19000],ac[5500:19000],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
QGss,f=mlab.psd(shear[5500:19000],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
QGaa,f=mlab.psd(ac[5500:19000],NFFT=N/nblocks,Fs=1./dt,window=mlab.window_hanning,noverlap=N/nblocks/2)
QGrr=QGss-abs(QGsa)**2/QGaa

fig,ax=plt.subplots(2,1)
fig.set_size_inches(9,6)

ax[0].loglog(f,TGaa,label='Accelerometer (turbulent)')
ax[0].loglog(f,TGss,label='Shear (turbulent)')
ax[0].loglog(f,TGrr,label='Vibration removed')
ax[0].set_title('Spectra of %d - %ds region'%(time[30000],time[N-1]))
ax[0].legend(loc='lower left')
ax[0].set_ylabel('G(f)')

ax[1].loglog(f,QGaa,label='Accelerometer (quiet)')
ax[1].loglog(f,QGss,label='Shear (quiet)')
ax[1].loglog(f,QGrr,label='Vibration removed')
ax[1].set_title('Spectra of %d - %ds region'%(time[5500],time[19000]))
ax[1].set_xlabel('f (Hz)')
ax[1].legend(loc='lower left')
ax[1].set_xlabel('f [Hz]')
ax[1].set_ylabel('G(f)')
```

When there is a lot of turbulence, the shear spectrum amplitude is much higher than it is in the "quiet" region. The real turbulence starts to overwhelm the vibration induced "turbulence" in the low frequency region of the "turbulent" spectrum, resulting in the 10-20Hz peak almost dissappearing. When there is very little true turbulence, the vibrations start to dominate, making the 10-20Hz peak more prominent.

The corrected turbulence spectra excludes the peak between 10 and 20 Hz, but still shows some of the peak at 50 Hz. The corrected spectrum is also much noisier in the "quiet" region, since there is effectively less data from natural turbulence.
