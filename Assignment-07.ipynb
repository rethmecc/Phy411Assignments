### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm

N=1001
dt=1.
t=np.arange(0,N*dt,dt)

x = np.cumsum(random.randn(N)) # red noise
f0=0.03 # Hz
x = x+ 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000.
x[N/2:]=x[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
x = x-np.mean(x)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
def prev():
    #1------------------------------
    fc=0.06
    X = np.fft.fft(xn)
    nfft=100
    ff = np.linspace(0,1./dt,N)
    X[(ff>fc) & (ff<(1./dt-fc))]=1.e-15
    y = np.real(np.fft.ifft(X))+np.mean(xn)
    args = {'NFFT':nfft,'Fs':1./dt,'noverlap':nfft/2.,'window':mlab.window_hanning}
    py,fff = mlab.psd(y,**args)
    xn0=xn*1.0
    x0=x*1.0
    fig,ax=plt.subplots(3,2,figsize=(20,15))
    ax[0][0].plot(t,x,label='signal')
    ax[0][0].plot(t,xn,label='signal+noise')
    ax[0][0].plot(t,y,label='smoothed')
    ax[0][0].legend()
    ax[0][0].set_xlabel('time (s)')
    ax[0][1].plot(t,x,label='signal')
    ax[0][1].plot(t,xn,label='signal+noise')
    ax[0][1].plot(t,y,label='smoothed')
    ax[0][1].set_xlim(400,600)
    ax[0][1].legend()
    ax[0][1].set_xlabel('time (s)')

    #2------------------------------
    Gxy,f  = mlab.csd(xn,y,**args)
    Gxnxn,f  = mlab.psd(xn,**args)
    Gyy,f  = mlab.psd(y,**args)
    Gxx,f  = mlab.psd(x,**args)
    Gnn,f  = mlab.psd(noise,**args)

    ax[1][0].loglog(f,np.sqrt(Gyy/(Gxnxn)),'k',label='boxcar')
    ax[1][0].set_ylabel(r'$|H(f)|$')
    ax[1][0].set_xlabel('f [Hz]')

    #3------------------------------         
    ax[1][1].semilogx(f,np.angle(Gxy),'k',label='boxcar')

    ax[1][1].set_ylabel('phase')
    ax[1][1].set_xlabel('f [Hz]')
    
    ax[2][0].plot(t,x,label='signal')
    ax[2][0].plot(t,xn,label='signal+noise')
    ax[2][0].plot(t,y,label='smoothed')
    ax[2][0].legend(loc='lower left')
    ax[2][0].set_xlim(300,500)
    ax[2][0].set_ylim(500,1500)
    ax[2][0].set_xlabel('time (s)')
    
    ax[2][1].plot(t,x,label='signal')
    ax[2][1].plot(t,xn,label='signal+noise')
    ax[2][1].plot(t,y,label='smoothed')
    ax[2][1].legend(loc='lower left')
    ax[2][1].set_xlim(100,300)
    ax[2][1].set_ylim(500,1500)
    ax[2][1].set_xlabel('time (s)')
    
    return fig,ax
prev()
```

It's good for the parts of the signal that are not close to the jump. Near the jump, it is not as good, but overall it is much better than before.

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python
def part2(fc=0.06,ran=(0,1)):
    fig,ax=prev()
    #1------------------------------
    X = np.fft.fft(xn)
    nfft=100
    ff = np.linspace(0,1./dt,N)

    ax[0][1].cla()
    ax[0][1].plot(ff,X,'.',label='before')
    
    X[(ff>fc) & (ff<(1./dt-fc))]=1.e-15
    len1=len(X[ff<fc])
    len2=len(X[ff>(1./dt-fc)])
    
    X[(ff<fc)] = X[(ff<fc)]*np.hanning(len1*2)[len1:]
    X[ff>(1./dt-fc)]=X[ff>(1./dt-fc)]*np.hanning(len2*2)[:len2]
    
    ax[0][1].plot(ff,X,label='windowed')
    ax[0][1].set_xlim(ran)
    ax[0][1].legend()
    ax[0][1].set_xlabel('f [Hz]')
    ax[0][1].set_title('FFT of signal')
    y = np.real(np.fft.ifft(X))
    args = {'NFFT':nfft,'Fs':1./dt,'noverlap':nfft/2.,'window':mlab.window_hanning}
    py,fff = mlab.psd(y,**args)
    xn0=xn*1.0
    x0=x*1.0
    ax[0][0].plot(t,y,label='with hanning',color='m')
    ax[2][0].plot(t,y,label='with hanning',color='m')
    ax[2][1].plot(t,y,label='with hanning',color='m')
    ax[0][0].legend()
    ax[2][0].legend()
    ax[2][1].legend()
    ax[0][0].set_xlim(300,700)

    #2------------------------------
    Gxy,f  = mlab.csd(xn,y,**args)
    Gxnxn,f  = mlab.psd(xn,**args)
    Gyy,f  = mlab.psd(y,**args)
    Gxx,f  = mlab.psd(x,**args)
    Gnn,f  = mlab.psd(noise,**args)

    ax[1][0].loglog(f,np.sqrt(Gyy/(Gxnxn)),'b',label='Hanning')
    ax[1][0].set_ylabel(r'$|H(f)|$')
    ax[1][0].set_ylim(1e-6,1.1)
    ax[1][0].legend()
    ax[1][0].plot((fc,fc),(1e-6,1),'k--')
    ax[1][0].text(fc,0.001,'$ \ \ \ \ \ f_c=%1.2f$' %fc,fontsize=12)
    ax[1][0].plot((0.03,0.03),(1e-6,1),'k--')
    ax[1][0].text(0.03,0.001,'$ \ \ \ \ \ 0.03Hz$',fontsize=12)

    #3------------------------------         
    ax[1][1].semilogx(f,np.angle(Gxy),'b',label='Hanning')

    ax[1][1].set_ylabel('phase')
    ax[1][1].set_xlabel('f [Hz]')
    ax[1][1].legend()
    
    return fig,ax
part2()
```

This filter does a better job with the sine wave, and doesn't do too bad with the jump. It falls off a little earlier, though.

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
fig2,ax2=part2(fc=0.1500001,ran=(0,0.1))
fig2.suptitle('fc = 0.15Hz',fontsize=30)
fig3,ax3=part2(fc=0.06,ran=(0,0.1))
fig3.suptitle('fc = 0.06Hz',fontsize=30)
```

## Q2: Implementing a bandpass filter

<div style='background:#F0F0F0'>**1** Create a red-noise process with 20000 data points, $dt=1\ s$.  Using a **truncated Dirichlet Kernel**, create a **bandpass** filter centered at $f_c=0.1\ Hz$ that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz.

   1. Plot the unshifted and shifted filters versus $k$, the time index.

HINT: when you make the Dirichlet Kernel filter make sure that the $k=0$ case is treated "correctly" because $sin(0)=0$ and you will get a discontinuity.  Calculate what $h_0$ should be from [L'Hôpital's_rule](http://en.wikipedia.org/wiki/L'Hôpital's_rule).

```python
N=20000
dt=1.
x = np.cumsum(np.random.randn(N))
print x
X=np.fft.fft(x)
ff = np.linspace(0,1./dt,N)
print ff
y=x*0.
y=X*0.
fc=0.1/2.

M = fc*2*N+1
k = np.array(range(-N/2+1,N/2+1))
h2 = (1./N)*np.sin(np.pi*(k)*M/N)/np.sin(np.pi*(k)/N)
h2[np.where(k==0)] = (1./N)*(np.pi*M/N)/(np.pi/N)

H1 = h2*0+1.
fig,ax=plt.subplots(2,2,figsize=(20,10))
nblocks=10
nfft=N/nblocks
args = {'NFFT':nfft,'Fs':1./dt,'noverlap':nfft/2.,'window':mlab.window_hanning}

K=1001
h = h2
h[abs(k)>(K-1)/2]=0
w = 0.*h
w[abs(k)<=(K-1)/2]=np.hanning(K)
h=h*w
H = N*np.fft.fftshift(np.fft.ifft(h))
Y=np.fft.fftshift(H*X)
y=h*x
ysh=y*2*np.cos(2*0.1*np.pi*k)

Gyy,f  = mlab.psd(y,**args)
Gxy,f  = mlab.csd(x,y,**args)
Gyysh,f  = mlab.psd(ysh,**args)
Gxysh,f  = mlab.csd(x,ysh,**args)
print nfft

ax[0][0].plot(k,y,label='Power spectrum before shift')
ax[0][0].plot(k,ysh,label='Power spectrum after shift')
ax[0][0].legend()
ax[0][0].set_ylabel(r'$h_k$');ax[0][0].set_xlabel(r'time: $k$')
ax[0][0].set_xlim(-100,100)

ax[0][1].loglog(f,Gyy,label='Filter in time space before shift')
ax[0][1].loglog(f,Gyysh,label='Filter in time space after shift')
ax[0][1].set_ylabel('G_{yy}')
ax[0][1].set_xlabel('f [Hz]')
ax[0][1].tick_params(axis='x', labelsize=13.5)
ax[0][1].legend(loc='lower left')

ax[1][0].plot(k,abs(H**2),label='unshifted')
ax[1][0].set_title('Filter in frequency space')
Hsh=N*np.fft.fftshift(np.fft.ifft(ysh))
ax[1][0].plot(k-0.1*N,abs(H**2),label='shifted',color='g');ax[1][0].plot(k+0.1*N,abs(H**2),label='',color='g')
ax[1][0].legend()
ax[1][0].set_ylabel('$H^2(m)$')
ax[1][0].set_xlabel('frequency m')

ax[1][1].semilogx(f,np.angle(Gxy),label='unshifted')
ax[1][1].semilogx(f,np.angle(Gxysh),label='shifted')
ax[1][1].legend()
ax[1][1].set_xlim(0.04,0.3)
ax[1][1].set_xlabel('f [Hz]')
ax[1][1].set_ylabel('phase')
ax[1][1].tick_params(axis='x', labelsize=13.5)

plt.tight_layout()
```

From the spectrum it can be seen that the filter does indeed attenuate the high and low frequencies, and passes the 0.1Hz frequency with essentially 0 attenuation.
