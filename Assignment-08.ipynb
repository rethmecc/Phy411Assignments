### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
N=1000.
x=np.random.randn(N)
dt=0.1
t=np.arange(0,N*dt,dt)

B=0.5

fp=B*0.7
fs = B*0.9
fNyq=1/2./dt
n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 60.)
b,a=signal.ellip(n,.1,60.,fn)
xBL = signal.filtfilt(b,a,x)

fig,ax=plt.subplots(2,1,figsize=(5,5))

ax[0].plot(t,x,color='0.5',label='full-spectrum signal')
ax[0].plot(t,xBL,color='k',label='band-limited signal')
ax[0].set_xlim([0,30]);ax[0].set_xlabel('time [s]');ax[0].set_ylabel('x [V]');ax[0].legend(fontsize='small',loc=0)

nfft=200
pxx,f=mlab.psd(x,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
pxBL,f=mlab.psd(xBL,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
ax[1].loglog(f,pxx,color='0.5')
ax[1].loglog(f,pxBL,color='k')
plt.axvline(x=B,linestyle='--',color='k');ax[1].text(B,1,' B=0.5 Hz');ax[1].set_ylim([1e-8,1e2]);ax[1].set_xlabel('f [Hz]');ax[1].set_ylabel('$G_{xx} [V^2 Hz^{-1}]$')
plt.tight_layout()
```

<div style='background:#F0F0F0'>**2** Now decimate by a factor of 4 to get a time series sampled at 2.5 Hz.  Show that you can recover the 10 Hz time series using the Sampling Theorem. </div>

```python
d=4
fig,axs=plt.subplots(2,2,figsize=(10,6))
ax=axs[0][0]
ax.plot(t,xBL,color='k',label='band-limited signal')
ax.plot(t[::d],xBL[::d],'d',color='b',label='Sampled at 2.5Hz')
ax.set_xlim([0,4]);ax.set_xlabel('time [s]');ax.set_ylabel('x [V]');ax.legend(fontsize='small',loc=0)
ax=axs[1][0]
ax.plot(t,xBL,color='k',label='band-limited signal')
ax.plot(t[::d],xBL[::d],color='b',label='Sampled at 2.5Hz')
ax.set_xlim([0,4]);ax.set_xlabel('time [s]');ax.set_ylabel('x [V]');ax.legend(fontsize='small',loc=0)
ax=axs[0][1]
ax.plot(t,xBL,color='k',label='band-limited signal')
ax.plot(t[::d],xBL[::d],'d',color='b',label='Sampled at 2.5Hz')
ax.set_xlim([0,30]);ax.set_xlabel('time [s]');ax.set_ylabel('x [V]');ax.legend(fontsize='small',loc=0)
ax=axs[1][1]
ax.plot(t,xBL,color='k',label='band-limited signal')
ax.plot(t[::d],xBL[::d],color='b',label='Sampled at 2.5Hz')
ax.set_xlim([0,30]);ax.set_xlabel('time [s]');ax.set_ylabel('x [V]');ax.legend(fontsize='small',loc=0)
```

Sampling at 2.5 Hz completely reproduces the bandwidth limited signal from the 10Hz time series. This makes sense, since $\frac{1}{\Delta t} = 2.5 > 2B = 1$.

```python
N=6
M=np.zeros((N+1,N+1))
X=np.zeros(N+1)
M[0][0]=1;M[N][N]=1

for m in range(1,N):
    M[m,m-1:m+2]=(t[m]-t[m-1],2*(t[m+1]-t[m-1]),t[m+1]-t[m])
    X[m]=((x[m+1]-x[m])/(t[m+1]-t[m])-(x[m]-x[m-1])/(t[m]-t[m-1]))*3.

inverse=np.linalg.inv(M)
c = np.dot(inverse,X)

d=np.zeros(N)
for i in range(0,N-1):
    d[i]=(c[i+1]-c[i])/3./(t[i+1]-t[i])
d[N-1]=-c[N-1]/3./(t[N]-t[N-1])

b=np.zeros(N)

for i in range(0,N):
    b[i]=(x[i+1]-x[i]-(t[i+1]-t[i])**2*c[i]-(t[i+1]-t[i])**3*d[i])/(t[i+1]-t[i])

a=x[:N]

q=np.zeros((N,N+1))
t = np.array([0.,2.5,2.7,3.4,4.2,6.5,9.0])
def q(T):
    i = np.where(T<=t[1:])[0][0]
    return a[i]+b[i]*(T-t[i])+c[i]*(T-t[i])**2+d[i]*(T-t[i])**3

time=np.linspace(t[0],t[N],100)
Q=np.zeros(100)

for ind in range(len(time)):
    Q[ind]=q(time[ind])

x0= np.cos(time*2*np.pi/3)

f = interpolate.interp1d(t, x,'cubic')

xnew = f(time)

fig,ax=plt.subplots(1,1)
ax.plot(t, x, 'o',label='Samples')
ax.plot(time, xnew,'-',label='Canned Fit')
ax.plot(time,x0,'.',label='Truth data')
ax.plot(time,Q,'-',label='Manual Fit')
ax.set_xlabel('time (s)')
ax.set_ylabel('x')
ax.legend(loc=0)
```
