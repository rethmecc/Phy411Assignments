## Q1: Lag correlations of weather time series

````python
import matplotlib.pyplot as plt
import numpy as np
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb = hourdata[1,:]
````

**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xy}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememebr to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
import numpy as np
import matplotlib.pyplot as plt

good=np.isfinite(dc+jb)
x=dc[good]
y=jb[good]

days = 30
lags =np.arange(days*24+1)
cxx=0.*lags
  
for ind,tau in enumerate(lags):
    if tau>0:
       cxx[ind]=np.mean((x[:-tau]-np.mean(x[:-tau]))*(x[tau:]-np.mean(x[tau:])))   
    elif tau==0:
       cxx[ind]=np.mean((x-np.mean(x))**2)
              
rowxx=cxx/np.std(x)/np.std(x)


fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(rowxx)
ax.set_xlabel(r'$\tau [hr]$',fontsize='18')
ax.set_ylabel(r'$\rho_{xx}$',fontsize='18')

````

With the increase of $\tau$, $\rho_{xx}$ decrease with a regular fuctuation. Every 24 hours,it will experience a period. when comparing with daytime and nighttime data, $\rho_{xx}$ will go up.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
#winter time: Dec.,Jan.,Feb.
#summer time:Jue.,Jul.,Aug.
windc=dc[11*24*30:14*24*30]
sumdc=dc[5*24*30:8*24*30]

windc=windc[np.isfinite(windc)]
sumdc=sumdc[np.isfinite(sumdc)]

x1=windc;
y1=sumdc;

cxy1=0.*lags;cxy2=0.*lags;
  
for ind,tau in enumerate(lags):
    if tau>0:
       cxy1[ind]=np.mean((x1[:-tau]-np.mean(x1[:-tau]))*(x1[tau:]-np.mean(x1[tau:])))    
       cxy2[ind]=np.mean((y1[:-tau]-np.mean(y1[:-tau]))*(y1[tau:]-np.mean(y1[tau:]))) 
    elif tau==0:
       cxy1[ind]=np.mean((x1-np.mean(x1))*(x1-np.mean(x1)))
       cxy2[ind]=np.mean((y1-np.mean(y1))*(y1-np.mean(y1))) 
    
                         
    
rowxy1=cxy1/np.std(x1)/np.std(x1)
rowxy2=cxy2/np.std(y1)/np.std(y1)

fig,ax = plt.subplots(2,1,figsize=(10,10))
ax[0].plot(rowxy1,label='winter months')
ax[0].set_xlabel(r'$\tau [hr]$',fontsize='18')
ax[0].set_ylabel(r'$\rho_{xy}$',fontsize='18')
ax[0].set_title('lag correlation between dc and jb temperature')
ax[0].legend()
ax[1].plot(rowxy2,'r',label='summer months')
ax[1].set_xlabel(r'$\tau [hr]$',fontsize='18')
ax[1].set_ylabel(r'$\rho_{xy}$',fontsize='18')
ax[1].set_title('lag correlation between dc and jb temperature')
ax[1].legend()
````

As for the summer months, $\rho_{xy}$ oscillates regularly in daytime and nightime. In the whole three months(summer months), the temperature has a lower frequecy of about 200hours.
````python
def lagautcor(N,b,lenlags):
    barray = np.ones(b)
    y=np.random.randn(N)
    x=np.convolve(y,barray,mode='same')
    lags = np.arange(lenlags)
    Rxx=0.*lags
    
    for ind,tau in enumerate(lags):
       if tau>0:
          Rxx[ind]=np.mean((x[:-tau])*(x[tau:]))   
       elif tau==0:
          Rxx[ind]=np.mean(x*x)
    return Rxx

N=[500,5000,50000,500000]
b=50
lenlags=100;
Rxx=np.zeros((len(N),lenlags))
             
for ind,Nn in enumerate(N):
    Rxx[ind]=lagautcor(Nn,b,lenlags)


fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(Rxx[0],label='N=%d' %N[0])
ax.plot(Rxx[1],label='N=%d' %N[1])
ax.plot(Rxx[2],label='N=%d' %N[2])
ax.plot(Rxx[3],label='N=%d' %N[3])
ax.set_xlabel(r'$\tau [hr]$',fontsize='18')
ax.set_ylabel(r'$R_{xx}$',fontsize='18')
ax.legend()       
````
