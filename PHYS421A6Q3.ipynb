```python
#import libraries
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.signal as signal
from scipy.interpolate import interp1d
from datetime import date
import scipy.interpolate as interpolate
import matplotlib.dates as dates
import PIL
import Image
import os
```

```python
#define function to simulate magnet
def run(T=2.,flips=50000,init=np.ones((50,50))):
    #initialize spins
    spins=np.ones((50,50))*init
    #random particle locations
    rx=np.random.randint(0,50,flips)
    ry=np.random.randint(0,50,flips)
    #define BJ
    BJ=1/(T)
    #magnetic moment array
    m=np.zeros(flips)
    #cycle through spins
    for i in range(flips):
        #apply periodic boundary conditions
        left=rx[i]-1
        right=rx[i]+1
        up=ry[i]-1
        down=ry[i]+1
        if left<0:
            left=49
        if right>49:
            right=0
        if up<0:
            up=49
        if down>49:
            down=0
        #calculate energy difference from spin flip
        dE=2*(spins[rx[i],ry[i]]*spins[right,ry[i]]
            +spins[rx[i],ry[i]]*spins[left,ry[i]]
            +spins[rx[i],ry[i]]*spins[rx[i],down]
            +spins[rx[i],ry[i]]*spins[rx[i],up])
        #Apply Metropolis algorithm
        if dE<=0:
            spins[rx[i],ry[i]]=spins[rx[i],ry[i]]*-1
        else:
            p=random.uniform(0,1)
            if np.exp(-BJ*dE)>p:
                spins[rx[i],ry[i]]=spins[rx[i],ry[i]]*-1
        #calculate magnetic moment
        m[i]=np.sum(spins)/2500.
    #calculate mean magnetic moment
    mBar=np.mean(m)
    #find equilibrium time
    eq=-1
    for n in range(flips):
        if eq==-1 and m[n]<mBar:
            eq=n
    return mBar,eq
```

```python
#number of temperature values
numT=40
Ts=np.linspace(0.1,6,numT)
mBars=np.zeros(numT)
time=np.zeros(numT)

#do runs
for rs in range(numT):
    mBars[rs],time[rs]=run(Ts[rs],500000)
print mBars

#make plots
fig,ax=plt.subplots(1,2,figsize=(15,5))
ax[0].plot(Ts,mBars)
ax[1].plot(Ts,time)
```
