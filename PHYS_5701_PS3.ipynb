```python
#My usual Python import list...
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import scipy.signal as signal
from scipy.interpolate import interp1d
from datetime import date
import scipy.interpolate as interpolate
import matplotlib.dates as dates
import PIL
import Image
import os
import math as math
from sympy.mpmath import mp
import scipy.special as sp
import numpy.lib.scimath as smath
import cmath as cmath
from scipy.integrate import quad
from scipy.optimize import minimize
```

```python
#Mathematical Functions used
def phase_shift(a,V0,l):
    q=smath.sqrt(1.-2.*V0) #define parameter 'q'
    numerator=sp.sph_jn(n=l,z=a)[1][l]-q*sp.sph_jn(n=l,z=q*a)[1][l]*sp.sph_jn(n=l,z=a)[0][l]/sp.sph_jn(n=l,z=q*a)[0][l]
    denominator=sp.sph_yn(n=l,z=a)[1][l]-q*sp.sph_jn(n=l,z=q*a)[1][l]*sp.sph_yn(n=l,z=a)[0][l]/sp.sph_jn(z=q*a,n=l)[0][l]
    delta_l=math.atan(numerator/denominator)
    return delta_l

def partial_amp(a,V0,l):
    delta_l=phase_shift(a,V0,l) #Get phase shift
    f_l=(2.*l+1)*cmath.exp(1j*delta_l)*math.sin(delta_l) #Calculate partial wave amplitude from phase shift
    return f_l

def scat_amp(a,V0,mu,n):
    f_l=np.zeros((n+1,),dtype=complex) #Create empty array of complex numbers
    for l in range(0,n+1):
        f_l[l]=partial_amp(a,V0,l) #Get partial wave amplitudes up to and including l=n
    if np.isscalar(mu): #If mu is one point
        f=np.sum(f_l*(sp.lpn(n=n,z=mu)[0])) #Sum of fl*Pl
    else: #If mu is array of points
        f=np.zeros(len(mu)) #Create empty array for f(mu)
        for i in range(len(mu)):
            f[i]=np.sum(f_l*(sp.lpn(n=n,z=mu[i])[0])) #Calculate sum of fl*Pl for each value of mu
    return f

def dcs(a,V0,mu,n): #Theoretical differential cross-section
    f=scat_amp(a,V0,mu,n) #Get amplitude
    return np.abs(f)**2 #Modulus squared

def obs(mu): #Observed (Yukawa) cross-section
    return (1./(17./9.-mu))**2

def Chi2_integrand(x,a,V0,n): #Integrand for Chi^2
    return abs(obs(x)-dcs(a,V0,x,n))**2

def Chi2(x,n): #Chi^2; done seperately from integral to allow use of built in integration function and minimizer
    return quad(Chi2_integrand, -1, 1, args=(x[0],x[1],n))[0]
```

```python
#Plot of cross sections
a=1.7841055
V0=0.37732921
n=50

mu=np.linspace(-1,1,10000) #Array of 10000 evenly spaced mu's
fig,ax=plt.subplots(1,2,figsize=(20,6))

#Plot of comparison of "observed" cross-section with best fit
ax[0].plot(mu,obs(mu),'--',label='observed')
ax[0].plot(mu,dcs(a,V0,mu,n),label='theory')
ax[0].legend(loc=2)
ax[0].set_ylabel(r'$\frac{d\sigma}{d\Omega}$',size=25)
ax[0].set_xlabel(r'$\cos(\theta)$',size=20)
ax[0].set_title(r'Figure 1: Observed and Predicted Differential Cross-Sections',size=18)

#Plot of "fun" cross-secton
a=20.
V0=15.
n=50
ax[1].plot(mu,dcs(a,V0,mu,n),label='Fun')
ax[1].set_ylabel(r'$\frac{d\sigma}{d\Omega}$',size=25)
ax[1].set_xlabel(r'$\cos(\theta)$',size=20)
ax[1].set_title(r'Figure 2: Spherical Cross-Section with $k=1,\/a=20,\/V_0=15$',size=18)
```

```python
#Optimization with 25 partial waves (a and V0 have been tuned by inputting the results of previous optimizations as the initial
#parameters of the next optimization)
bnds = ((0, None), (None, None))
print minimize(Chi2,[1.7841055,0.37732921],bounds=bnds,method='Nelder-Mead',args=(25,))
```

```python
#Opimization with 50 partial waves
bnds = ((0, None), (None, None))
print minimize(Chi2,[1.7841055,0.37732921],bounds=bnds,method='Nelder-Mead',args=(50,))
```

The optimal value of $a$ is 1.7841055

The optimal value of $V_0$ is 0.37732921

The resulting $\chi^2$ is 0.0029669783686057643

The optimization was not affected when I went from 25 partial waves to 50.
