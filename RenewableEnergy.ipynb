```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
Full=np.genfromtxt('RE/Full.txt')
DarkBlue=np.genfromtxt('RE/DarkBlue.txt')
DarkMagenta=np.genfromtxt('RE/DarkMagenta.txt')
DarkOrange=np.genfromtxt('RE/DarkOrange.txt')
DarkPink=np.genfromtxt('RE/DarkPink.txt')
DarkPurple=np.genfromtxt('RE/DarkPurple.txt')
DarkYellow=np.genfromtxt('RE/DarkYellow.txt')
Green=np.genfromtxt('RE/Green.txt')
LightBlue=np.genfromtxt('RE/LightBlue.txt')
LightMagenta=np.genfromtxt('RE/LightMagenta.txt')
LightOrange=np.genfromtxt('RE/LightOrange.txt')
LightPink=np.genfromtxt('RE/LightPink.txt')
LightPurple=np.genfromtxt('RE/LightPurple.txt')
LightYellow=np.genfromtxt('RE/LightYellow.txt')
White=np.genfromtxt('RE/White.txt')

fig,ax=plt.subplots(5,3)
fig.set_size_inches(20,15)
ax[0][0].plot(Full[:,0],Full[:,1],label='Full')
ax[0][1].plot(DarkBlue[:,0],DarkBlue[:,1],label='DarkBlue')
ax[0][2].plot(DarkMagenta[:,0],DarkMagenta[:,1],label='DarkMagenta')
ax[1][0].plot(DarkOrange[:,0],DarkOrange[:,1],label='DarkOrange')
ax[1][1].plot(DarkPink[:,0],DarkPink[:,1],label='DarkPink')
ax[1][2].plot(DarkPurple[:,0],DarkPurple[:,1],label='DarkPurple')
ax[2][0].plot(DarkYellow[:,0],DarkYellow[:,1],label='DarkYellow')
ax[2][1].plot(Green[:,0],Green[:,1],label='Green')
ax[2][2].plot(LightBlue[:,0],LightBlue[:,1],label='LightBlue')
ax[3][0].plot(LightMagenta[:,0],LightMagenta[:,1],label='LightMagenta')
ax[3][1].plot(LightOrange[:,0],LightOrange[:,1],label='LightOrange')
ax[3][2].plot(LightPink[:,0],LightPink[:,1],label='LightPink')
ax[4][0].plot(LightPurple[:,0],LightPurple[:,1],label='LightPurple')
ax[4][1].plot(LightYellow[:,0],LightYellow[:,1],label='LightYellow')
ax[4][2].plot(White[:,0],White[:,1],label='White')
for i in range(5):
    for j in range(3):
        ax[i][j].legend()
```
